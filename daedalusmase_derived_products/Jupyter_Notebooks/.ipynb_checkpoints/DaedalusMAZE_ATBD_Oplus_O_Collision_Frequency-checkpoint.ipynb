```python
#import required packages
import sys
sys.path.insert(1, "../SourceCode")
import DaedalusGlobals as DaedalusGlobals
from netCDF4 import Dataset
import numpy as np
import datetime
import pyglow
from ipywidgets import*
import ipywidgets as widgets
import warnings
import glob
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from mpl_toolkits.basemap import Basemap, shiftgrid, addcyclic
import matplotlib.mlab as mlab
from mpl_toolkits.axes_grid1 import make_axes_locatable
from scipy.interpolate import griddata
from pylab import *
from matplotlib.patches import Polygon
import time
import pandas as pd
import image
from scipy.special import erf
import matplotlib.colors as colors
from matplotlib import ticker, cm
from scipy.interpolate import make_interp_spline, BSpline
from apexpy import Apex
#Allocations

#Allocations
zg1=[]
nO1=[]
nO21=[]
nN21=[]
Ne1=[]
NOp1=[]
NO2p1=[]
sigmaHall1=[]
sigmaPed1=[]
sigma01=[]
vOp1=[]
ven1=[]
vNOp1=[]
NNOp1=[]
vO2p1=[]
vin1=[]
T_e=[]
T_i=[]
T_n=[]
T_i_iri=[]
T_n_iri=[]
lat_out=[]
T_e_iri=[]
lon_out=[]
Argon=[]
Helium=[]
NitricOxide=[]
Nitrogen_ion=[]
vei=[]
sigmaPed_Op1=[]
sigmaPed_O2p1=[]
sigmaPed_NOp1=[]
sigmaPed_e1=[]

sigmaHall_Op1=[]
sigmaHall_O2p1=[]
sigmaHall_NOp1=[]
sigmaHall_e1=[]
maptime=np.zeros((24),order='F')

Tr_out=[]
q_dalgarno_out=[]
q_banks_out=[]
q_stubbe_out=[]
q_sw_out=[]
q_salah_out=[]
q_pesnell_out=[]
q_hickmann_out=[]
q_shunk_nagy_out=[]
q_richmond_out=[]

Tr_iri=[]
q_dalgarno_iri=[]
q_banks_iri=[]
q_stubbe_iri=[]
q_sw_iri=[]
q_salah_iri=[]
q_pesnell_iri=[]
q_hickmann_iri=[]
q_shunk_nagy_iri=[]
q_richmond_iri=[]

sigmaPed_O2p_Ο21=[]
sigmaPed_O2p_Ο1=[]
sigmaPed_O2p_N21=[]
sigmaPed_Op_Ο21=[]
sigmaPed_Op_Ο1=[]
sigmaPed_Op_N21=[]
sigmaPed_NOp_Ο21=[]
sigmaPed_NOp_Ο1=[]
sigmaPed_NOp_N21=[]

sigmaHall_O2p_Ο21=[]
sigmaHall_O2p_Ο1=[]
sigmaHall_O2p_N21=[]
sigmaHall_Op_Ο21=[]
sigmaHall_Op_Ο1=[]
sigmaHall_Op_N21=[]
sigmaHall_NOp_Ο21=[]
sigmaHall_NOp_Ο1=[]
sigmaHall_NOp_N21=[]

sigmaPed_f08_out=[]
sigmaPed_f09_out=[]
sigmaPed_f10_out=[]
sigmaPed_f11_out=[]
sigmaPed_f12_out=[]
sigmaPed_f13_out=[]
sigmaPed_f14_out=[]
sigmaPed_f15_out=[]
sigmaPed_f16_out=[]
sigmaPed_f17_out=[]
sigmaPed_f18_out=[]

sigmaHall_f08_out=[]
sigmaHall_f09_out=[]
sigmaHall_f10_out=[]
sigmaHall_f11_out=[]
sigmaHall_f12_out=[]
sigmaHall_f13_out=[]
sigmaHall_f14_out=[]
sigmaHall_f15_out=[]
sigmaHall_f16_out=[]
sigmaHall_f17_out=[]
sigmaHall_f18_out=[]
                    
    
#Transformation from ENU to ECEF
def enu_ecef(lat_phi, lon_lmd, Fe, Fn, Fup):
    fac = np.pi / 180
    lat_phi = lat_phi * fac
    lon_lmd = lon_lmd * fac

    north_temp_unit = [-np.cos(lon_lmd) * np.sin(lat_phi), - np.sin(lon_lmd) * np.sin(lat_phi), np.cos(lat_phi)]
    east_temp_unit = [-np.sin(lon_lmd), np.cos(lon_lmd), 0]
    up_temp_unit = [np.cos(lon_lmd) * np.cos(lat_phi), np.sin(lon_lmd) * np.cos(lat_phi), np.sin(lat_phi)]

    Fnorth_vector = [Fn * north_temp_unit[0], Fn * north_temp_unit[1], Fn * north_temp_unit[2]]
    Feast_vector = [Fe * east_temp_unit[0], Fe * east_temp_unit[1], Fe * east_temp_unit[2]]
    Fup_vector = [Fup * up_temp_unit[0], Fup * up_temp_unit[1], Fup * up_temp_unit[2]]

    Fx = Fnorth_vector[0] + Feast_vector[0] + Fup_vector[0]
    Fy = Fnorth_vector[1] + Feast_vector[1] + Fup_vector[1]
    Fz = Fnorth_vector[2] + Feast_vector[2] + Fup_vector[2]

    return Fx, Fy, Fz

#Transformation from ECEF to ENU
def ecef_enu(lat_phi, lon_lmd, Fx, Fy, Fz):
    fac = np.pi / 180
    lat_phi = lat_phi * fac
    lon_lmd = lon_lmd * fac
    
    x_temp_unit=[-np.sin(lon_lmd), -np.cos(lon_lmd)*np.sin(lat_phi),np.cos(lon_lmd)*np.cos(lat_phi)]
    y_temp_unit=[np.cos(lon_lmd),-np.sin(lon_lmd)*np.sin(lat_phi),np.sin(lon_lmd)*np.cos(lat_phi)]
    z_temp_unit=[0, np.cos(lat_phi),np.sin(lat_phi)]

    Fx_vector = [Fx * x_temp_unit[0], Fx * x_temp_unit[1], Fx * x_temp_unit[2]]
    Fy_vector = [Fy * y_temp_unit[0], Fy * y_temp_unit[1], Fy * y_temp_unit[2]]
    Fz_vector = [Fz * z_temp_unit[0], Fz * z_temp_unit[1], Fz * z_temp_unit[2]]

    Fe = Fx_vector[0] + Fy_vector[0] + Fz_vector[0]
    Fn = Fx_vector[1] + Fy_vector[1] + Fz_vector[1]
    Fu = Fx_vector[2] + Fy_vector[2] + Fz_vector[2]

    return Fe, Fn, Fu

 # calculate geocentric latitude from geodetic latitude (WGS84)
def geo_lat2geod_lat(phi):
    a = 6378137  # meter semi major axis of earth
    f = 1 / 298.257  # flattening
    b = a - f * a  # semi minor axis
    e = ((a ** 2 - b ** 2) ** (1 / 2)) / a
    phi_rad = np.deg2rad(phi)
    geod_lat = np.arctan(np.tan(phi_rad) / (1 - e ** 2))
    geod_lat = np.rad2deg(geod_lat)
    return geod_lat  # in degrees

#Geographic to Geodetic transformation
def geo2geod(lat_geo_phi, lon_geo_lmd, alt_geo):
    lat_geod_phi = geo_lat2geod_lat(lat_geo_phi)  # degrees
    lon_geod_lmd = lon_geo_lmd  # degrees
    alt_geod = alt_geo  # km
    return lat_geod_phi, lon_geod_lmd, alt_geod 

def collisions():
    Tr=np.arange(500,2000,10)
    for i in range(0,len(Tr)):
        q_dalgarno=4.3*10**(-11)*Tr[i]**(0.41)
        q_banks=3.5*10**(-11)*np.sqrt(Tr[i])*((1-np.log10(Tr[i]))**2)
        q_stubbe=7.2*10**(-11)*(Tr[i]**0.37)
        
        q_sw=3.69*10**(-11)*np.sqrt(Tr[i])*((1-np.log10(Tr[i]))**2)
        q_salah=4*10**(-11)*np.sqrt(Tr[i])
        q_pesnell=5.9*10**(-11)*np.sqrt(Tr[i])*((1-0.096*np.log10(Tr[i]))**2)
        q_hickmann=5.92*10**(-11)*(Tr[i]**(0.393))*(1+(96.6/Tr[i])**2)
        q_shunk_nagy=3.67 * (10**(-11)) * np.sqrt(Tr[i])*((1-0.064*np.log10(Tr[i]))**2)
        q_richmond=6.7 * (10**(-11)) * np.sqrt(Tr[i])*((0.96-0.135*np.log10(Tr[i]))**2)
        
        Tr_out.append(Tr[i])
        q_dalgarno_out.append(q_dalgarno)
        q_banks_out.append(q_banks)
        q_stubbe_out.append(q_stubbe)
        q_sw_out.append(q_sw)
        q_salah_out.append(q_salah)
        q_pesnell_out.append(q_pesnell)
        q_hickmann_out.append(q_hickmann)
        q_shunk_nagy_out.append(q_shunk_nagy)
        q_richmond_out.append(q_richmond)
#     print(q_stubbe_out)
   
def run(tiegcm_file,lat_value,lon_value,timer_value):
    # Initializations
    start_time = time.time()
    time_igrf = datetime.datetime(2015, 3, 15, 0, 0, 0)  # first time step of the TIEGCM run
    timer = 0  # counter used for incrementing TIEGCM timestep
    deltalmd = 2.5  # TIEGCM grid resolution in deg
    deltaphi = 2.5  # TIEGCM grid resolution in deg
    Re = 6378.1370  # Earths Radius (km)
    electron = 1.602176565 * 10 ** (- 19) #electron charge in C
    boltzmann_si=1.38064852 * 10**(-23)
    boltzmann=1.380645852 * 10 ** (-16) #Boltzmann constant in cm^2*g*s^(-2)*K^(-1)
    me=9.10938356 * 10 ** (- 31) #electron mass in kg
    mO=16 # Oxygen atomic mass in g/mol
    mN2=28  #N2 molecular mass in g/mol
    mO2=32 #O2 molecular mass in g/mol
    mNO=30 #NO molecular mass in g/mol
    NA=6.02214086 * 10 ** 23 #Avogadro's constant in mol^-1
    fcor=1.5 #Burnside facor (default=1.5)
    #TIEGCM = Dataset(DaedalusGlobals.TIEGCM_Files_Path +"TIEGCM_EVT1_2015_StPatricksDay_HAO/tiegcm_dres.s_mar2015_amie_v1_31.nc")
    TIEGCM = Dataset(tiegcm_file)
    # Get data from TIEGCM Model file
    glat = TIEGCM.variables['lat'][:] #geographic latitude in deg
    glon = TIEGCM.variables['lon'][:] #geographic longitude in deg
    glev = TIEGCM.variables['lev'][:] #interface levels
    gtime = TIEGCM.variables['time'][:] #time in minutes since 2015 March 15 00:00:00
    zg = TIEGCM.variables['ZGMID'][:] #Geometric height in cm
    Ne_all = TIEGCM.variables['NE'][:]  # electron density in cm^-3
    Ne_mid = TIEGCM.variables['ELECDEN'][:]  # electron density in cm^-3
    PED_all = TIEGCM.variables['PEDERSEN'][:] #Pedersen conductivity in Si/m
    HALL_all= TIEGCM.variables['HALL'][:] #Hall conductivity in Si/m
    Un_east = TIEGCM.variables['UN'][:]  # neutral zonal wind (+East) in cm/s
    Un_north = TIEGCM.variables['VN'][:]  # neutral meridional wind (+North) in cm/s
    Un_up = TIEGCM.variables['WN_lev'][:]  # neutral vertical wind (+Up) in cm/s
    Ui_east = TIEGCM.variables['Ui_lev'][:]  # zonal ExB velocity in cm/s
    Ui_north = TIEGCM.variables['Vi_lev'][:]  # meridional ExB velocity in cm/s
    Ui_up = TIEGCM.variables['Wi_lev'][:]  # vertical ExB velocity in cm/s
    Poten_all=TIEGCM.variables['POTEN'][:] #electric potential in V
    barm_all=TIEGCM.variables['BARM'][:] #Mean molecular weight in g/mol
    P0=TIEGCM.variables['p0'][:] #Reference pressure in millibar
    Zh=TIEGCM.variables['Z'][:] #dimensionless variable (for pressure calculations)
    Tn_all=TIEGCM.variables['TN'][:] #Neutral Temperature in K
    Te_all=TIEGCM.variables['TE'][:] #Electron Temperature in K
    Ti_all=TIEGCM.variables['TI'][:] #Ion Temperature in K
    nO_all=TIEGCM.variables['O1'][:] #neutral atomic oxygen density in mmr
    nO2_all=TIEGCM.variables['O2'][:] #neutral molecular oxygen density in mmr
    NOp_all=TIEGCM.variables['OP'][:] #O+ density in cm^-3
    den_all=TIEGCM.variables['DEN'][:] #Density
    nO_all_cm3=TIEGCM.variables['O_CM3'][:] #neutral atomic oxygen density in mmr
    nO2_all_cm3=TIEGCM.variables['O2_CM3'][:] #neutral molecular oxygen density in mmr
    NO2p_all=TIEGCM.variables['O2P'][:] #O+ density in cm^-3
    Np_all=TIEGCM.variables['NPLUS'][:] #N+ density in cm^-3
    NNOp_all=TIEGCM.variables['NOP_LAM'][:] #NO+ density in cm^-3
    N2_all_cm3=TIEGCM.variables['N2_CM3'][:] #O+ density in cm^-3
    EEX=TIEGCM.variables['EEX'][:] 
    EEY=TIEGCM.variables['EEY'][:] 
    EEZ=TIEGCM.variables['EEZ'][:] 
    QAMIE=TIEGCM.variables['QAMIE'][:] 
    QWIND=TIEGCM.variables['QWIND'][:]
    Argon_all=TIEGCM.variables['AR'][:]
    Helium_all=TIEGCM.variables['HE'][:]
    NitricOxide_all=TIEGCM.variables['NO'][:]
    Nitrogen_ion_all=TIEGCM.variables['NPLUS'][:]
    xnmbar_all=TIEGCM.variables['XNMBAR'][:]
    
    
    TIEGCM.close() #close input file
    



    #for timer in range(0, len(gtime)):
    for timer in range(timer_value,timer_value+1):
        time_plot=gtime[timer]
        maptime[timer]=time_plot

        for lev in range(0, len(glev)):
        #for lev in range(53, 54):
            
            #for lat in range(0, len(glat)):
            for lat in range(lat_value, lat_value+1):

                #for lon in range(0, len(glon)):
                for lon in range(lon_value, lon_value+1):
                    
                    ########################################################################
#                   Coordinates
                    ########################################################################
                    
                    alt_p = zg[timer, lev, lat, lon] / 1e5 #cm to km
                    lat_p=glat[lat]
                    lon_p = glon[lon]
                    lat_out.append(lat_p)
                    lon_out.append(lon_p)
                    zg1.append(alt_p)
                    timeg=gtime[timer]
                    #time
                    time_p = time_igrf + datetime.timedelta(minutes=timeg)
                    Z=glev[lev]
                    ########################################################################
#                   Temperatures and densities
                    ########################################################################

                    #Temperatures
                    Tn=Tn_all[timer, lev, lat, lon]     #Kelvin
                    Te=Te_all[timer, lev, lat, lon]     #Kelvin
                    Ti=Ti_all[timer, lev, lat, lon]     #Kelvin
                    
                    T_e.append(Te)
                    T_i.append(Ti)
                    T_n.append(Tn)

                    #mean molecular mass g/mol
                    barm=barm_all[timer, lev, lat, lon]
                    
                    #neutral Densities (cm^-3)
                    NO=nO_all_cm3[timer, lev, lat, lon] #cm^-3
                    NO2=nO2_all_cm3[timer, lev, lat, lon] #cm^-3                 
                    NN2=N2_all_cm3[timer, lev, lat, lon] #cm^-3
                    
                    nO1.append(NO)
                    nO21.append(NO2)
                    nN21.append(NN2)
                    
                    # electron density
                    Ne = Ne_mid[timer, lev, lat, lon] * (10 ** 6)  # m^-3
                    Necm=Ne_mid[timer, lev, lat, lon] 
                    Ne1.append(Necm)

                    #ion Densities
                    NOp=NOp_all[timer, lev, lat, lon]  #cm^-3
                    NO2p=NO2p_all[timer, lev, lat, lon]  #cm^-3
                    NOp1.append(NOp)
                    NO2p1.append(NO2p)
                    NNOp=NNOp_all[timer, lev, lat, lon]   #if O2+ density is available in the input file
                    NNOp1.append(NNOp) #output 

                    MO=mO/(NA*1000)  # Atomic oxygen mass in kg
                    MO2=(2*mO)/(NA*1000) # Molecular oxygen mass in kg
                    MNO=(mNO)/(NA*1000) #Nitric oxide mass in kg
                    MHe=4/(NA*1000)
                    MN=14/(NA*1000)
                    
                    #Conversion factor (mmr->cm^(-3))
                    Nbarm=(P0*np.exp(-Z)*barm*1000)/(boltzmann*Tn)
                    
                    Ar_argon=39.948
                    XNMBAR_f=xnmbar_all[timer, lev, lat, lon]
                    ARG=Argon_all[timer, lev, lat, lon]
                    HEL=Helium_all[timer, lev, lat, lon]
                    NITRIC=NitricOxide_all[timer, lev, lat, lon]
                    argon_f=(XNMBAR_f*ARG)/Ar_argon
                    Argon.append(argon_f)
                    Ar_helium=4.002
                    helium_f=(XNMBAR_f*HEL)/Ar_helium
                    Helium.append(helium_f)
                    Ar_NO=30
                    NO_f=(XNMBAR_f*NITRIC)/Ar_NO
                    NitricOxide.append(NO_f)
                    Nplus=Nitrogen_ion_all[timer, lev, lat, lon]
                    Nitrogen_ion.append(Nplus)
                    
                    
                    ########################################################################
#                   Magnetic and Electric fields
                    ########################################################################
                    
                    # Magnetic Field from ENU to ECEF
                    pt = pyglow.Point(time_p, lat_p, lon_p, alt_p)
                    pt.run_igrf()
                    Be = pt.Bx
                    Bn = pt.By
                    Bu = pt.Bz
                    Bx, By, Bz = enu_ecef(lat_p, lon_p, Be, Bn, Bu)  # Tesla in ECEF
                    B = [Bx, By, Bz]  # vector of magnetic field ECEF in Telsa
                    
                    
                    pt.run_iri()
                    Ti_iri=pt.Ti
                    Tn_iri=pt.Tn_iri
                    Te_iri=pt.Te
                    T_i_iri.append(Ti_iri)
                    T_n_iri.append(Tn_iri)
                    T_e_iri.append(Te_iri)
                       
                    # Magnetic field unit vector
                    bnorm = np.sqrt(Bx * Bx + By * By + Bz * Bz)
                    b_unit = [Bx / bnorm, By / bnorm, Bz / bnorm]



                    ########################################################################
#                   Gyrofrequencies and collision frequencies
                    ########################################################################
                   
                    #gyrofrequencies (Hz)
                    bgauss=bnorm*10000
                    qeomeo10=1.7588028*10**7
                    qeoNa010=9.6489*10**3
                    rmassinv_op=1/mO
                    rmassinv_nop=1/mNO
                    rmassinv_o2p=1/mO2
                    
                    omega_op=qeoNa010*bgauss*rmassinv_op
                    omega_o2p=qeoNa010*bgauss*rmassinv_o2p
                    omega_nop=qeoNa010*bgauss*rmassinv_nop
                    omega_op_inv=1/omega_op
                    omega_o2p_inv=1/omega_o2p
                    omega_nop_inv=1/omega_nop
                    omega_e=qeomeo10*bgauss
                    omega_e_inv=1/omega_e

                    #O+_O freqs
                    Tr=(Ti_iri+Tn_iri)/2
                    
                    q_dalgarno=4.3*10**(-11)*Tr**(0.41)
                    q_banks=3.5*10**(-11)*np.sqrt(Tr)*((1-np.log10(Tr))**2)
                    q_stubbe=7.2*10**(-11)*(Tr**0.37)

                    q_sw=3.69*10**(-11)*np.sqrt(Tr)*((1-np.log10(Tr))**2)
                    q_salah=4*10**(-11)*np.sqrt(Tr)
                    q_pesnell=5.9*10**(-11)*np.sqrt(Tr)*((1-0.096*np.log10(Tr))**2)
                    q_hickmann=5.92*10**(-11)*(Tr**(0.393))*(1+(96.6/Tr)**2)
                    q_shunk_nagy=3.67 * (10**(-11)) * np.sqrt(Tr)*((1-0.064*np.log10(Tr))**2)
                    q_richmond=6.7 * (10**(-11)) * np.sqrt(Tr)*((0.96-0.135*np.log10(Tr))**2)

                    Tr_iri.append(Tr)
                    q_dalgarno_iri.append(q_dalgarno)
                    q_banks_iri.append(q_banks)
                    q_stubbe_iri.append(q_stubbe)
                    q_sw_iri.append(q_sw)
                    q_salah_iri.append(q_salah)
                    q_pesnell_iri.append(q_pesnell)
                    q_hickmann_iri.append(q_hickmann)
                    q_shunk_nagy_iri.append(q_shunk_nagy)
                    q_richmond_iri.append(q_richmond)
                    
                    
                    
                    
                    
                    #collision frequencies (Hz)
                    temps=(Ti+Tn)/2
                    
                    rnu_o2p_o2=2.59 * (10**(-11))* np.sqrt(temps)*((1-0.073*np.log10(temps))**2)
                    rnu_op_o2=6.64 * (10**(-10))
                    rnu_nop_o2=4.27 * (10**(-10))
                    rnu_np_o2=7.25 * (10**(-10))
                    
                    rnu_o2p_o=2.31*(10**(-10))
                    rnu_op_o=3.67 * (10**(-11)) * np.sqrt(temps)*((1-0.064*np.log10(temps))**2)*fcor

                    rnu_op_o_f_0_8=3.67 * (10**(-11)) * np.sqrt(temps)*((1-0.064*np.log10(temps))**2)*0.8
                    rnu_op_o_f_0_9=3.67 * (10**(-11)) * np.sqrt(temps)*((1-0.064*np.log10(temps))**2)*0.9
                    rnu_op_o_f_1_0=3.67 * (10**(-11)) * np.sqrt(temps)*((1-0.064*np.log10(temps))**2)
                    rnu_op_o_f_1_1=3.67 * (10**(-11)) * np.sqrt(temps)*((1-0.064*np.log10(temps))**2)*1.1
                    rnu_op_o_f_1_2=3.67 * (10**(-11)) * np.sqrt(temps)*((1-0.064*np.log10(temps))**2)*1.2
                    rnu_op_o_f_1_3=3.67 * (10**(-11)) * np.sqrt(temps)*((1-0.064*np.log10(temps))**2)*1.3
                    rnu_op_o_f_1_4=3.67 * (10**(-11)) * np.sqrt(temps)*((1-0.064*np.log10(temps))**2)*1.4
                    rnu_op_o_f_1_5=3.67 * (10**(-11)) * np.sqrt(temps)*((1-0.064*np.log10(temps))**2)*1.5
                    rnu_op_o_f_1_6=3.67 * (10**(-11)) * np.sqrt(temps)*((1-0.064*np.log10(temps))**2)*1.6
                    rnu_op_o_f_1_7=3.67 * (10**(-11)) * np.sqrt(temps)*((1-0.064*np.log10(temps))**2)*1.7
                    rnu_op_o_f_1_8=3.67 * (10**(-11)) * np.sqrt(temps)*((1-0.064*np.log10(temps))**2)*1.8
                    
                    rnu_nop_o=2.44 * (10**(-10))
                    rnu_np_o=4.42 * (10**(-10))
                    
                    rnu_o2p_n2=4.13*(10**(-10))
                    rnu_op_n2=6.82 * (10**(-10))
                    rnu_nop_n2=4.34*(10**(-10))
                    rnu_np_n2=7.47*(10**(-10))
                    
                    rnu_o2p_he=0.7* (10**(-10))
                    rnu_op_he=1.32* (10**(-10))
                    rnu_nop_he=0.74* (10**(-10))
                    rnu_np_he=1.49*(10**(-10))
                    
                    rnu_o2p=(rnu_o2p_o2*NO2+rnu_o2p_o*NO+rnu_o2p_n2*NN2)*omega_o2p_inv
                    rnu_op=(rnu_op_o2*NO2+rnu_op_o*NO+rnu_op_n2*NN2)*omega_op_inv
                    
                    rnu_op_f_0_8=(rnu_op_o2*NO2+rnu_op_o_f_0_8*NO+rnu_op_n2*NN2)*omega_op_inv
                    rnu_op_f_0_9=(rnu_op_o2*NO2+rnu_op_o_f_0_9*NO+rnu_op_n2*NN2)*omega_op_inv
                    rnu_op_f_1_0=(rnu_op_o2*NO2+rnu_op_o_f_1_0*NO+rnu_op_n2*NN2)*omega_op_inv
                    rnu_op_f_1_1=(rnu_op_o2*NO2+rnu_op_o_f_1_1*NO+rnu_op_n2*NN2)*omega_op_inv
                    rnu_op_f_1_2=(rnu_op_o2*NO2+rnu_op_o_f_1_2*NO+rnu_op_n2*NN2)*omega_op_inv
                    rnu_op_f_1_3=(rnu_op_o2*NO2+rnu_op_o_f_1_3*NO+rnu_op_n2*NN2)*omega_op_inv
                    rnu_op_f_1_4=(rnu_op_o2*NO2+rnu_op_o_f_1_4*NO+rnu_op_n2*NN2)*omega_op_inv
                    rnu_op_f_1_5=(rnu_op_o2*NO2+rnu_op_o_f_1_5*NO+rnu_op_n2*NN2)*omega_op_inv
                    rnu_op_f_1_6=(rnu_op_o2*NO2+rnu_op_o_f_1_6*NO+rnu_op_n2*NN2)*omega_op_inv
                    rnu_op_f_1_7=(rnu_op_o2*NO2+rnu_op_o_f_1_7*NO+rnu_op_n2*NN2)*omega_op_inv
                    rnu_op_f_1_8=(rnu_op_o2*NO2+rnu_op_o_f_1_8*NO+rnu_op_n2*NN2)*omega_op_inv
                    
                    rnu_nop=(rnu_nop_o2*NO2+rnu_nop_o*NO+rnu_nop_n2*NN2)*omega_nop_inv
                    
                    r_o2p_o2=rnu_o2p_o2*NO2*omega_o2p_inv
                    r_o2p_o=rnu_o2p_o*NO*omega_o2p_inv
                    r_o2p_n2=rnu_o2p_n2*NN2*omega_o2p_inv
                    r_op_o2=rnu_op_o2*NO2*omega_op_inv
                    r_op_o=rnu_op_o*NO*omega_op_inv
                    r_op_n2=rnu_op_n2*NN2*omega_op_inv
                    r_nop_o2=rnu_nop_o2*NO2*omega_nop_inv
                    r_nop_o=rnu_nop_o*NO*omega_nop_inv
                    r_nop_n2=rnu_nop_n2*NN2*omega_nop_inv
                    
                    
                    vOp=rnu_op_o2*NO2+rnu_op_o*NO+rnu_op_n2*NN2
                    vO2p=rnu_o2p_o2*NO2+rnu_o2p_o*NO+rnu_o2p_n2*NN2
                    vNOp=rnu_nop_o2*NO2+rnu_nop_o*NO+rnu_nop_n2*NN2
                    vin=vOp+vO2p+vNOp
                    
                   
                    vop_o=rnu_op_o*NO
                    vop_o2=rnu_op_o2*NO2
                    vop_n2=rnu_op_n2*NN2
                    vo2p_o=rnu_o2p_o*NO
                    vo2p_o2=rnu_o2p_o2*NO2
                    vo2p_n2=rnu_o2p_n2*NN2
                    vnop_o=rnu_nop_o*NO
                    vnop_o2=rnu_nop_o2*NO2
                    vnop_n2=rnu_nop_n2*NN2
                    vop_he=rnu_op_he*helium_f
                    vo2p_he=rnu_o2p_he*helium_f
                    vnop_he=rnu_nop_he*helium_f
                    vnp_o=rnu_np_o*NO
                    vnp_o2=rnu_np_o2*NO2
                    vnp_n2=rnu_np_n2*NN2
                    vnp_he=rnu_np_he*helium_f
                    
                    
                    ven=(2.33 * (10**(-11))*NN2*Te*(1-(1.21*(10**(-4)*Te))) \
                        + 1.82 * (10**(-10))*NO2*np.sqrt(Te)*(1+(3.6*(10**(-2)*np.sqrt(Te)))) \
                        +8.9 * (10**(-11))*NO*np.sqrt(Te)*(1+(5.7*(10**(-4)*Te))))
                    
                    ve_n2=2.33 * (10**(-11))*NN2*Te*(1-(1.21*(10**(-4)*Te)))
                    ve_o=8.9 * (10**(-11))*NO*np.sqrt(Te)*(1+(5.7*(10**(-4)*Te)))
                    ve_o2=1.82 * (10**(-10))*NO2*np.sqrt(Te)*(1+(3.6*(10**(-2)*np.sqrt(Te))))
                    ve_he=4.6* (10**(-10))*helium_f*np.sqrt(Te)
                    
                    rnu_ne=(2.33 * (10**(-11))*NN2*Te*(1-(1.21*(10**(-4)*Te))) \
                        + 1.82 * (10**(-10))*NO2*np.sqrt(Te)*(1+(3.6*(10**(-2)*np.sqrt(Te)))) \
                        +8.9 * (10**(-11))*NO*np.sqrt(Te)*(1+(5.7*(10**(-4)*Te))))*omega_e_inv
                    rnu_ne=4*rnu_ne
                    
                    vOp1.append(vOp)
                    vO2p1.append(vO2p)
                    vNOp1.append(vNOp)
                    vin1.append(vin)
                    
                    #electron-ion (Coulomb collisions)
                    
                    ve_op=(54.5*NOp)/(Te**(1.5))
                    ve_o2p=(54.5*NO2p)/(Te**(1.5))
                    ve_nop=(54.5*NNOp)/(Te**(1.5))
                    ve_np=(54.5*Nplus)/(Te**(1.5))
                    ve_i=ve_op+ve_o2p+ve_nop+ve_np
                    vei.append(ve_i)
                    

                    
                    ########################################################################
#                   Conductivities
                    ########################################################################                
                    
                    qe_fac=(electron*10**10)/bgauss
                    Necm2=NOp+NO2p+NNOp
                    sigmaPed=qe_fac*((NOp*rnu_op/(1+rnu_op**2))+(NO2p*rnu_o2p/(1+rnu_o2p**2)) \
                                    +(NNOp*rnu_nop/(1+rnu_nop**2))+Necm2*rnu_ne/(1+rnu_ne**2))
                    
                    sigmaHall=qe_fac*(Necm2/(1+rnu_ne**2)-NOp/(1+rnu_op**2)-NO2p/(1+rnu_o2p**2)-NNOp/(1+rnu_nop**2))
                    
                    #different fcor
                    sigmaPed_f_0_8=qe_fac*((NOp*rnu_op_f_0_8/(1+rnu_op_f_0_8**2))+(NO2p*rnu_o2p/(1+rnu_o2p**2)) \
                            +(NNOp*rnu_nop/(1+rnu_nop**2))+Necm2*rnu_ne/(1+rnu_ne**2))
                    sigmaPed_f_0_9=qe_fac*((NOp*rnu_op_f_0_9/(1+rnu_op_f_0_9**2))+(NO2p*rnu_o2p/(1+rnu_o2p**2)) \
                            +(NNOp*rnu_nop/(1+rnu_nop**2))+Necm2*rnu_ne/(1+rnu_ne**2))
                    sigmaPed_f_1_0=qe_fac*((NOp*rnu_op_f_1_0/(1+rnu_op_f_1_0**2))+(NO2p*rnu_o2p/(1+rnu_o2p**2)) \
                            +(NNOp*rnu_nop/(1+rnu_nop**2))+Necm2*rnu_ne/(1+rnu_ne**2))    
                    sigmaPed_f_1_1=qe_fac*((NOp*rnu_op_f_1_1/(1+rnu_op_f_1_1**2))+(NO2p*rnu_o2p/(1+rnu_o2p**2)) \
                            +(NNOp*rnu_nop/(1+rnu_nop**2))+Necm2*rnu_ne/(1+rnu_ne**2))
                    sigmaPed_f_1_2=qe_fac*((NOp*rnu_op_f_1_2/(1+rnu_op_f_1_2**2))+(NO2p*rnu_o2p/(1+rnu_o2p**2)) \
                            +(NNOp*rnu_nop/(1+rnu_nop**2))+Necm2*rnu_ne/(1+rnu_ne**2))
                    sigmaPed_f_1_3=qe_fac*((NOp*rnu_op_f_1_3/(1+rnu_op_f_1_3**2))+(NO2p*rnu_o2p/(1+rnu_o2p**2)) \
                            +(NNOp*rnu_nop/(1+rnu_nop**2))+Necm2*rnu_ne/(1+rnu_ne**2))
                    sigmaPed_f_1_4=qe_fac*((NOp*rnu_op_f_1_4/(1+rnu_op_f_1_4**2))+(NO2p*rnu_o2p/(1+rnu_o2p**2)) \
                            +(NNOp*rnu_nop/(1+rnu_nop**2))+Necm2*rnu_ne/(1+rnu_ne**2))
                    sigmaPed_f_1_5=qe_fac*((NOp*rnu_op_f_1_5/(1+rnu_op_f_1_5**2))+(NO2p*rnu_o2p/(1+rnu_o2p**2)) \
                            +(NNOp*rnu_nop/(1+rnu_nop**2))+Necm2*rnu_ne/(1+rnu_ne**2))
                    sigmaPed_f_1_6=qe_fac*((NOp*rnu_op_f_1_6/(1+rnu_op_f_1_6**2))+(NO2p*rnu_o2p/(1+rnu_o2p**2)) \
                            +(NNOp*rnu_nop/(1+rnu_nop**2))+Necm2*rnu_ne/(1+rnu_ne**2))
                    sigmaPed_f_1_7=qe_fac*((NOp*rnu_op_f_1_7/(1+rnu_op_f_1_7**2))+(NO2p*rnu_o2p/(1+rnu_o2p**2)) \
                            +(NNOp*rnu_nop/(1+rnu_nop**2))+Necm2*rnu_ne/(1+rnu_ne**2))
                    sigmaPed_f_1_8=qe_fac*((NOp*rnu_op_f_1_8/(1+rnu_op_f_1_8**2))+(NO2p*rnu_o2p/(1+rnu_o2p**2)) \
                            +(NNOp*rnu_nop/(1+rnu_nop**2))+Necm2*rnu_ne/(1+rnu_ne**2))
                    
                    sigmaHall_f_0_8=qe_fac*(Necm2/(1+rnu_ne**2)-NOp/(1+rnu_op_f_0_8**2)-NO2p/(1+rnu_o2p**2)-NNOp/(1+rnu_nop**2))
                    sigmaHall_f_0_9=qe_fac*(Necm2/(1+rnu_ne**2)-NOp/(1+rnu_op_f_0_9**2)-NO2p/(1+rnu_o2p**2)-NNOp/(1+rnu_nop**2))
                    sigmaHall_f_1_0=qe_fac*(Necm2/(1+rnu_ne**2)-NOp/(1+rnu_op_f_1_0**2)-NO2p/(1+rnu_o2p**2)-NNOp/(1+rnu_nop**2))
                    sigmaHall_f_1_1=qe_fac*(Necm2/(1+rnu_ne**2)-NOp/(1+rnu_op_f_1_1**2)-NO2p/(1+rnu_o2p**2)-NNOp/(1+rnu_nop**2))
                    sigmaHall_f_1_2=qe_fac*(Necm2/(1+rnu_ne**2)-NOp/(1+rnu_op_f_1_2**2)-NO2p/(1+rnu_o2p**2)-NNOp/(1+rnu_nop**2))
                    sigmaHall_f_1_3=qe_fac*(Necm2/(1+rnu_ne**2)-NOp/(1+rnu_op_f_1_3**2)-NO2p/(1+rnu_o2p**2)-NNOp/(1+rnu_nop**2))
                    sigmaHall_f_1_4=qe_fac*(Necm2/(1+rnu_ne**2)-NOp/(1+rnu_op_f_1_4**2)-NO2p/(1+rnu_o2p**2)-NNOp/(1+rnu_nop**2))
                    sigmaHall_f_1_5=qe_fac*(Necm2/(1+rnu_ne**2)-NOp/(1+rnu_op_f_1_5**2)-NO2p/(1+rnu_o2p**2)-NNOp/(1+rnu_nop**2))
                    sigmaHall_f_1_6=qe_fac*(Necm2/(1+rnu_ne**2)-NOp/(1+rnu_op_f_1_6**2)-NO2p/(1+rnu_o2p**2)-NNOp/(1+rnu_nop**2))
                    sigmaHall_f_1_7=qe_fac*(Necm2/(1+rnu_ne**2)-NOp/(1+rnu_op_f_1_7**2)-NO2p/(1+rnu_o2p**2)-NNOp/(1+rnu_nop**2))
                    sigmaHall_f_1_8=qe_fac*(Necm2/(1+rnu_ne**2)-NOp/(1+rnu_op_f_1_8**2)-NO2p/(1+rnu_o2p**2)-NNOp/(1+rnu_nop**2))
                    
                    sigmaPed_f08_out.append(sigmaPed_f_0_8)
                    sigmaPed_f09_out.append(sigmaPed_f_0_9)
                    sigmaPed_f10_out.append(sigmaPed_f_1_0)
                    sigmaPed_f11_out.append(sigmaPed_f_1_1)
                    sigmaPed_f12_out.append(sigmaPed_f_1_2)
                    sigmaPed_f13_out.append(sigmaPed_f_1_3)
                    sigmaPed_f14_out.append(sigmaPed_f_1_4)
                    sigmaPed_f15_out.append(sigmaPed_f_1_5)
                    sigmaPed_f16_out.append(sigmaPed_f_1_6)
                    sigmaPed_f17_out.append(sigmaPed_f_1_7)
                    sigmaPed_f18_out.append(sigmaPed_f_1_8)

                    sigmaHall_f08_out.append(sigmaHall_f_0_8)
                    sigmaHall_f09_out.append(sigmaHall_f_0_9)
                    sigmaHall_f10_out.append(sigmaHall_f_1_0)
                    sigmaHall_f11_out.append(sigmaHall_f_1_1)
                    sigmaHall_f12_out.append(sigmaHall_f_1_2)
                    sigmaHall_f13_out.append(sigmaHall_f_1_3)
                    sigmaHall_f14_out.append(sigmaHall_f_1_4)
                    sigmaHall_f15_out.append(sigmaHall_f_1_5)
                    sigmaHall_f16_out.append(sigmaHall_f_1_6)
                    sigmaHall_f17_out.append(sigmaHall_f_1_7)
                    sigmaHall_f18_out.append(sigmaHall_f_1_8)
                    
                    sigmaPed_Op=qe_fac*(NOp*rnu_op/(1+rnu_op**2))
                    sigmaPed_O2p=qe_fac*(NO2p*rnu_o2p/(1+rnu_o2p**2))
                    sigmaPed_NOp=qe_fac*(NNOp*rnu_nop/(1+rnu_nop**2))
                    sigmaPed_e=qe_fac*(Necm2*rnu_ne/(1+rnu_ne**2))
                    
                    sigmaPed_O2p_Ο2=qe_fac*(NO2p*r_o2p_o2/(1+r_o2p_o2**2))
                    sigmaPed_O2p_Ο=qe_fac*(NO2p*r_o2p_o/(1+r_o2p_o**2))
                    sigmaPed_O2p_N2=qe_fac*(NO2p*r_o2p_n2/(1+r_o2p_n2**2))
                    sigmaPed_Op_Ο2=qe_fac*(NOp*r_op_o2/(1+r_op_o2**2))
                    sigmaPed_Op_Ο=qe_fac*(NOp*r_op_o/(1+r_op_o**2))
                    sigmaPed_Op_N2=qe_fac*(NOp*r_op_n2/(1+r_op_n2**2))
                    sigmaPed_NOp_Ο2=qe_fac*(NNOp*r_nop_o2/(1+r_nop_o2**2))
                    sigmaPed_NOp_Ο=qe_fac*(NNOp*r_nop_o/(1+r_nop_o**2))
                    sigmaPed_NOp_N2=qe_fac*(NNOp*r_nop_n2/(1+r_nop_n2**2))
                    
                    sigmaHall_O2p_Ο2=-qe_fac*(NO2p/(1+r_o2p_o2**2))
                    sigmaHall_O2p_Ο=-qe_fac*(NO2p/(1+r_o2p_o**2))
                    sigmaHall_O2p_N2=-qe_fac*(NO2p/(1+r_o2p_n2**2))
                    sigmaHall_Op_Ο2=-qe_fac*(NOp/(1+r_op_o2**2))
                    sigmaHall_Op_Ο=-qe_fac*(NOp/(1+r_op_o**2))
                    sigmaHall_Op_N2=-qe_fac*(NOp/(1+r_op_n2**2))
                    sigmaHall_NOp_Ο2=-qe_fac*(NNOp/(1+r_nop_o2**2))
                    sigmaHall_NOp_Ο=-qe_fac*(NNOp/(1+r_nop_o**2))
                    sigmaHall_NOp_N2=-qe_fac*(NNOp/(1+r_nop_n2**2))
                    sigmaHall_Op=-qe_fac*(NOp/(1+rnu_op**2))
                    sigmaHall_O2p=-qe_fac*(NO2p/(1+rnu_o2p**2))
                    sigmaHall_NOp=-qe_fac*(NNOp/(1+rnu_nop**2))
                    sigmaHall_e=qe_fac*(Necm2/(1+rnu_ne**2))
                    
                    sigmaPed1.append(sigmaPed)

                    sigmaHall1.append(sigmaHall)

                    sigmaPed_Op1.append(sigmaPed_Op)
                    sigmaPed_O2p1.append(sigmaPed_O2p)
                    sigmaPed_NOp1.append(sigmaPed_NOp)
                    sigmaPed_e1.append(sigmaPed_e)

                    sigmaHall_Op1.append(sigmaHall_Op)
                    sigmaHall_O2p1.append(sigmaHall_O2p)
                    sigmaHall_NOp1.append(sigmaHall_NOp)
                    sigmaHall_e1.append(sigmaHall_e)
                   
                
                    sigmaPed_O2p_Ο21.append(sigmaPed_O2p_Ο2)
                    sigmaPed_O2p_Ο1.append(sigmaPed_O2p_Ο)
                    sigmaPed_O2p_N21.append(sigmaPed_O2p_N2)
                    sigmaPed_Op_Ο21.append(sigmaPed_Op_Ο2)
                    sigmaPed_Op_Ο1.append(sigmaPed_Op_Ο)
                    sigmaPed_Op_N21.append(sigmaPed_Op_N2)
                    sigmaPed_NOp_Ο21.append(sigmaPed_NOp_Ο2)
                    sigmaPed_NOp_Ο1.append(sigmaPed_NOp_Ο)
                    sigmaPed_NOp_N21.append(sigmaPed_NOp_N2)
                  
                    sigmaHall_O2p_Ο21.append(sigmaHall_O2p_Ο2)
                    sigmaHall_O2p_Ο1.append(sigmaHall_O2p_Ο)
                    sigmaHall_O2p_N21.append(sigmaHall_O2p_N2)
                    sigmaHall_Op_Ο21.append(sigmaHall_Op_Ο2)
                    sigmaHall_Op_Ο1.append(sigmaHall_Op_Ο)
                    sigmaHall_Op_N21.append(sigmaHall_Op_N2)
                    sigmaHall_NOp_Ο21.append(sigmaHall_NOp_Ο2)
                    sigmaHall_NOp_Ο1.append(sigmaHall_NOp_Ο)
                    sigmaHall_NOp_N21.append(sigmaHall_NOp_N2)
                    
    print("Calculation executed!")
    print("Total run time: %s seconds" % (time.time() - start_time))

    
warnings.simplefilter('ignore')

#window creation            
style = {'description_width': '150px'}
layout1stcolumn = {'width': '300px'}
layout2ndcolumn= {'width': '250px'}
layout3rdcolumn= {'width': '100px'}
button_layout={'width': '150px'}
style1 = {'description_width':'150px'}
layout1 = {'width':'850px'}
style2 = {'description_width':'95px'}
layout2 = {'width':'160px'}
layout3 = {'width':'350px'}
style3 = {'description_width':'60px'}
layout4 = {'width':'215px'}

tiegcm_file=widgets.Dropdown(options=sorted(glob.glob(DaedalusGlobals.TIEGCM_Files_Path +"TIEGCM_EVT1_2015_StPatricksDay_Rerun/tiegcm_dres.s_mar2015_amie_v1_*.nc")), description='TIE-GCM files: ', style=style1, layout=layout1)
orbit_file=widgets.Dropdown(options=sorted(glob.glob(DaedalusGlobals.Orbit_Files_Path +"DAED_ORB_Evt0_LLA*.csv")), description='Orbit files: ', style=style1, layout=layout1)
lat_value=widgets.Dropdown(
                        options=[('-88.75', 0), ('-86.25', 1), ('-83.75', 2),('-81.25', 3), ('-78.75', 4),('-76.25', 5), ('-73.75', 6),('-71.25', 7),
                        ('-68.75', 8), ('-66.25', 9), ('-63.75', 10),('-61.25', 11), ('-58.75', 12),('-56.25', 13), ('-53.75', 14),('-51.25', 15),
                        ('-48.75', 16), ('-46.25', 17), ('-43.75', 18),('-41.25', 19), ('-38.75', 20),('-36.25', 21), ('-33.75', 22),('-31.25', 23),
                        ('-28.75', 24), ('-26.25', 25), ('-23.75', 26),('-21.25', 27), ('-18.75', 28),('-16.25', 29), ('-13.75', 30),('-11.25', 31),
                        ('-8.75', 32), ('-6.25', 33), ('-3.75', 34),('-1.25', 35), ('1.25', 36),('3.75', 37), ('6.25', 38),('8.75', 39),
                        ('11.25', 40), ('13.75', 41), ('16.25', 42),('18.75', 43), ('21.25', 44),('23.75', 45), ('26.25', 46),('28.75', 47),
                        ('31.25', 48),('33.75', 49), ('36.25', 50),('38.75', 51), ('41.25', 52),('43.75', 53),
                        ('46.25', 54),('48.75', 55), ('51.25', 56),('53.75', 57), ('56.25', 58),('58.75', 59),
                        ('61.25', 60),('63.75', 61), ('66.25', 62),('68.75', 63), ('71.25', 64),('73.75', 65),
                        ('76.25', 66),('78.75', 67), ('81.25', 68),('83.75', 69), ('84.25', 70),('87.75', 71)],
                        value=61,
                        description='Latitude (deg):',
                        layout=layout1stcolumn, 
                        style=style
                        )
lon_value=widgets.Dropdown(
                        options=[('-180.0', 0), 
                                 ('-177.5', 1), ('-175.0', 2),('-172.5', 3), ('-170.0', 4),('-167.5', 5), ('-165.0', 6),('-162.5', 7), ('-160.0', 8), 
                                 ('-157.5', 9), ('-155.0', 10),('-152.5', 11), ('-150.0', 12),('-147.5', 13),('-145.0', 14),('-142.5', 15),('-140.0', 16), 
                                 ('-137.5', 17), ('-135.0', 18),('-132.5', 19), ('-130.0', 20),('-127.5', 21), ('-125.0', 22),('-122.5', 23),('-120.0', 24), 
                                 ('-117.5', 25), ('-115.0', 26),('-112.5', 27), ('-110.0', 28),('-107.5', 29), ('-105.0', 30),('-102.5', 31),('-100.0', 32), 
                                 ('-97.5', 33), ('-95.0', 34),('-92.5', 35),('-90.0', 36),('-87.5', 37),('-85.0', 38),('-82.5', 39),('-80.0', 40), 
                                 ('-77.5', 41), ('-75.0', 42),('-72.5', 43),('-70.0', 44),('-67.5', 45),('-65.0', 46),('-62.5', 47),('-60.0', 48),
                                 ('-57.5', 49), ('-55.0', 50),('-52.5', 51),('-50.0', 52),('-47.5', 53),('-45.0', 54),('-42.5', 55),('-40.0', 56),
                                 ('-37.5', 57), ('-35.0', 58),('-32.5', 59),('-30.0', 60),('-27.5', 61),('-25.0', 62),('-22.5', 63),('-20.0', 64),
                                 ('-17.5', 65), ('-15.0', 66),('-12.5', 67),('-10.0', 68),('-7.5', 69), ('-5.0', 70),('-2.5', 71),('0.0', 72),
                                 ('2.5', 73), ('5.0', 74),('7.5', 75),('10.0', 76),('12.5', 77), ('15.0', 78),('17.5', 79),('20.0', 80),('22.5', 81),
                                 ('25.0', 82), ('27.5', 83),('30.0', 84),('32.5', 85),('35.0', 86), ('37.5', 87),('40.0', 88),('42.5', 89),('45.0', 90),
                                 ('47.5', 91),('50.0', 92),('52.5', 93),('55.0', 94), ('57.5', 95),('60.0', 96),('62.5', 97),('65.0', 98),
                                 ('67.5', 99),('70.0', 100),('72.5', 101),('75.0', 102), ('77.5', 103),('80.0', 104),('82.5', 105),('85.0', 106),
                                 ('87.5', 107),('90.0', 108),('92.5', 109),('95.0', 110), ('97.5', 111),('100.0', 112),('102.5', 113),('105.0', 114),
                                 ('107.5', 115),('110.0', 116),('112.5', 117),('115.0', 118), ('117.5', 119),('120.0', 120),
                                 ('122.5', 121),('125.0', 122),('127.5', 123),('130.0', 124), ('132.5', 125),('135.0', 126),
                                 ('137.5', 127),('140.0', 128),('142.5', 129),('145.0', 130), ('147.5', 131),('150.0', 132),
                                 ('152.5', 133),('155.0', 134),('157.5', 135),('160.0', 136), ('162.5', 137),('165.0', 138),
                                 ('167.5', 139),('170.0', 140),('172.5', 141),('175.0', 142), ('177.5', 143),
                                ],
                        value=25,
                        description='Longitude (deg):',
                        layout=layout1stcolumn, 
                        style=style
                        )
plot_param=widgets.Dropdown(
            options=['Joule Heating','Frictional Heating','Ohmic Heating','Ohmic Heating per mass','Ohmic Heating per pressure',
                     'Pressure_Ohmic ratio', 'Convenction Heating', 'Ion Heating Rate',
                     'Wind Heating', 'Mechanical Power', 'Energy','Pedersen', 'Hall', 'Parallel','O','O2','N2', 'O+',
                     'O2+','NO+','Ne', 
                     'Ti','Tn','Te','νO+n','νO2+n','vNO+n','νin','νen', 'σ', 'σO+n','σO2+n','σNO+n','Potential','Pressure'],
            value='Joule Heating',
            description='Parameter to Plot:',
            disabled=False,
            style=style, 
            layout=layout1stcolumn
            )

quiver_back_param=widgets.Dropdown(
            options=['Potential', 'Joule Heating','Frictional Heating','Ohmic Heating','Convenction Heating','Ion Heating Rate',
                     'Wind Heating', 'Mechanical Power', 'Energy','Current','Pedersen','Hall','Parallel','Ion Drift'],
            value='Potential',
            description='Background:',
            disabled=False,
            style=style, 
            layout=layout1stcolumn
            )

quiver_vec_param=widgets.Dropdown(
            options=['Electric Field', 'Magnetic Field', 'Pedersen Current', 'Hall Current', 'Perpendicular Current', 'Ion Velocity',
                    'Neutral Winds','JxB'],
            value='Electric Field',
            description='Vector Field:',
            disabled=False,
            style=style, 
            layout=layout1stcolumn
            )
quiver_plot_param=widgets.Dropdown(
            options=['Vector','Streamplot','Divergence'],
            value='Vector',
            description='Projection:',
            disabled=False,
            style=style, 
            layout=layout1stcolumn
            )

timer_value= widgets.Dropdown( value=0, options=range(0,24),  description='TimeStep:', style=style, layout=layout1stcolumn)
pressure_level= widgets.Dropdown( value=0, options=range(0,57),  description='Pressure Level:', style=style, layout=layout1stcolumn)
heatings_checkbox   = widgets.Checkbox(value=True,  description="Plot Heating Rates", style=style1, layout=layout1 )
JHrho_checkbox   = widgets.Checkbox(value=False,  description="Plot Heating per mass", style=style1, layout=layout1 )
JHpress_checkbox   = widgets.Checkbox(value=False,  description="Plot Heating per pressure", style=style1, layout=layout1 )
pressure_heating_checkbox= widgets.Checkbox(value=False,  description="Plot Pressure/Heating ratio", style=style1, layout=layout1 )
pressure_heating_log_checkbox= widgets.Checkbox(value=False,  description="Plot Pressure/Heating ratio (log)", style=style1, layout=layout1 )
press_checkbox   = widgets.Checkbox(value=False,  description="Pressure", style=style1, layout=layout1 )
conductivities_checkbox   = widgets.Checkbox(value=False,  description="Plot Conductivities", style=style1, layout=layout1 )
frequencies_checkbox   = widgets.Checkbox(value=False,  description="Plot Frequencies", style=style1, layout=layout1 )
cross_checkbox   = widgets.Checkbox(value=False,  description="Plot Cross Sections", style=style1, layout=layout1 )
densities_checkbox   = widgets.Checkbox(value=False,  description="Plot Densities", style=style1, layout=layout1 )
temperatures_checkbox   = widgets.Checkbox(value=False,  description="Plot Temperatures", style=style1, layout=layout1 )
currents_checkbox   = widgets.Checkbox(value=False,  description="Plot Currents", style=style1, layout=layout1 )
efield_checkbox   = widgets.Checkbox(value=False,  description="Plot E field", style=style1, layout=layout1 )
velocities_checkbox   = widgets.Checkbox(value=False,  description="Plot Velocities", style=style1, layout=layout1 )
Joule_checkbox= widgets.Checkbox(value=True,  description="Map Plot",style=style1, layout=layout1 )
Joule_polar= widgets.Checkbox(value=True,  description="Polar Plot",style=style1, layout=layout1 )
Joule_ortho= widgets.Checkbox(value=True,  description="Orthographic Plot",style=style1, layout=layout3 )
Bounding_lat  = widgets.BoundedIntText(value=40, min=10, max=180, description='Bounding Latitude:', style=style1, layout=layout4)
North_plot= widgets.Checkbox(value=True,  description="Plot North Pole",style=style3, layout=layout3 )
South_plot= widgets.Checkbox(value=False,  description="Plot South Pole",style=style3, layout=layout3 )
center_lat  = widgets.BoundedIntText(value=45, min=-90, max=90, description='Center Latitude:', style=style1, layout=layout4)
center_lon  = widgets.BoundedIntText(value=45, min=-180, max=180, description='Center Longitude:', style=style1, layout=layout4)
quiver_checkbox= widgets.Checkbox(value=True,  description="Quiver Plot",style=style1, layout=layout1 )
orbit_checkbox= widgets.Checkbox(value=True,  description="Add Orbit",style=style1, layout=layout1 )
night_checkbox= widgets.Checkbox(value=True,  description="Add Nightshade",style=style1, layout=layout1)


def plot_conductivities(timer):
    
    time_begin=datetime.datetime(2015,3,15,0,0,0)
    time_plot=time_begin+datetime.timedelta(minutes=maptime[timer])
    
     
    fig3c, ax3c = plt.subplots(figsize=(10, 7))
#     plt.ticklabel_format(axis="x", style="sci", scilimits=(0,0))
    
    plt.suptitle("Temperatures")
    ax3c.set_title('%s Lattitude=%s Longitude=%s' % (time_plot.strftime("%d %b %Y %H:%M:%S"),lat_out[0],lon_out[0]))
#     ax1b.set_xscale('log')
    ax3c.plot(T_i[0:-1],zg1[0:-1],color='tab:blue', linewidth=2, label='Ti (TIEGCM)')
    ax3c.plot(T_n[0:-1],zg1[0:-1],color='tab:orange', linewidth=2, label='Tn (TIEGCM)')
    ax3c.plot(T_e[0:-1],zg1[0:-1],color='tab:pink', linewidth=2, label='Te (TIEGCM)')
    ax3c.plot(T_i_iri[0:-1],zg1[0:-1],color='tab:red', linewidth=2, label='Ti (IRI)')
    ax3c.plot(T_n_iri[0:-1],zg1[0:-1],color='tab:green', linewidth=2, label='Tn (IRI)')
    ax3c.plot(T_e_iri[0:-1],zg1[0:-1],color='tab:olive', linewidth=2, label='Te (IRI)')
#     ax3c.plot(Tr_out[0:-1],q_pesnell_out[0:-1],color='tab:pink', linewidth=2, label='Pesnell_1993')
#     ax3c.plot(Tr_out[0:-1],q_hickmann_out[0:-1],color='tab:olive', linewidth=2, label='Hickman_1997')
#     ax3c.plot(Tr_out[0:-1],q_shunk_nagy_out[0:-1],color='tab:gray', linewidth=2, label='Schun $ Nagy_2009')
#     ax3c.plot(Tr_out[0:-1],q_richmond_out[0:-1],color='tab:cyan', linewidth=2, label='Richmond_2006')
    ax3c.grid(True, color="#93a1a1", alpha=0.3)
    ax3c.minorticks_on()
    plt.legend(loc='best')
#     plt.ylim(100,400)
    ax3c.set_xlabel("Temperature (K)", labelpad=15, fontsize=12, color="#333533")
    ax3c.set_ylabel("Altitude (km)", labelpad=15, fontsize=12, color="#333533")

    plt.show()    
    fig3c, ax3c = plt.subplots(figsize=(10, 7))
#     plt.ticklabel_format(axis="x", style="sci", scilimits=(0,0))
    
    plt.suptitle("Hall Conductivity")
    ax3c.set_title('%s Lattitude=%s Longitude=%s' % (time_plot.strftime("%d %b %Y %H:%M:%S"),lat_out[0],lon_out[0]))
#     ax1b.set_xscale('log')
    ax3c.plot(sigmaHall1[0:-1],zg1[0:-1],color='tab:blue', linewidth=2, label='Hall')
    ax3c.plot(sigmaHall_Op1[0:-1],zg1[0:-1],color='tab:orange', linewidth=2, label='O+')
    ax3c.plot(sigmaHall_O2p1[0:-1],zg1[0:-1],color='tab:red', linewidth=2, label='O2+')
    ax3c.plot(sigmaHall_NOp1[0:-1],zg1[0:-1],color='tab:green', linewidth=2, label='NO+')
    ax3c.plot(sigmaHall_e1[0:-1],zg1[0:-1],color='tab:purple', linewidth=2, label='e-')
#     ax3c.plot(Tr_out[0:-1],q_pesnell_out[0:-1],color='tab:pink', linewidth=2, label='Pesnell_1993')
#     ax3c.plot(Tr_out[0:-1],q_hickmann_out[0:-1],color='tab:olive', linewidth=2, label='Hickman_1997')
#     ax3c.plot(Tr_out[0:-1],q_shunk_nagy_out[0:-1],color='tab:gray', linewidth=2, label='Schun $ Nagy_2009')
#     ax3c.plot(Tr_out[0:-1],q_richmond_out[0:-1],color='tab:cyan', linewidth=2, label='Richmond_2006')
    ax3c.grid(True, color="#93a1a1", alpha=0.3)
    ax3c.minorticks_on()
    plt.legend(loc='best')
#     plt.ylim(100,400)
    ax3c.set_xlabel("Conductivity (S/m)", labelpad=15, fontsize=12, color="#333533")
    ax3c.set_ylabel("Altitude (km)", labelpad=15, fontsize=12, color="#333533")

    plt.show()
    
    fig2c, ax2c = plt.subplots(figsize=(10, 7))
#     plt.ticklabel_format(axis="x", style="sci", scilimits=(0,0))
    plt.suptitle("Pedersen Conductivity")
    ax2c.set_title('%s Lattitude=%s Longitude=%s' % (time_plot.strftime("%d %b %Y %H:%M:%S"),lat_out[0],lon_out[0]))
#     ax1b.set_xscale('log')
    ax2c.plot(sigmaPed1[0:-1],zg1[0:-1],color='tab:blue', linewidth=2, label='Pedersen')
    ax2c.plot(sigmaPed_Op1[0:-1],zg1[0:-1],color='tab:orange', linewidth=2, label='O+')
    ax2c.plot(sigmaPed_O2p1[0:-1],zg1[0:-1],color='tab:red', linewidth=2, label='O2+')
    ax2c.plot(sigmaPed_NOp1[0:-1],zg1[0:-1],color='tab:green', linewidth=2, label='NO+')
    ax2c.plot(sigmaPed_e1[0:-1],zg1[0:-1],color='tab:purple', linewidth=2, label='e-')
#     ax2c.plot(Tr_out[0:-1],q_pesnell_out[0:-1],color='tab:pink', linewidth=2, label='Pesnell_1993')
#     ax2c.plot(Tr_out[0:-1],q_hickmann_out[0:-1],color='tab:olive', linewidth=2, label='Hickman_1997')
#     ax2c.plot(Tr_out[0:-1],q_shunk_nagy_out[0:-1],color='tab:gray', linewidth=2, label='Schun $ Nagy_2009')
#     ax2c.plot(Tr_out[0:-1],q_richmond_out[0:-1],color='tab:cyan', linewidth=2, label='Richmond_2006')
    ax2c.grid(True, color="#93a1a1", alpha=0.3)
    ax2c.minorticks_on()
    plt.legend(loc='best')
#     plt.ylim(100,400)
    ax2c.set_xlabel("Conductivity (S/m)", labelpad=15, fontsize=12, color="#333533")
    ax2c.set_ylabel("Altitude (km)", labelpad=15, fontsize=12, color="#333533")

    plt.show()
#################################################################################################

    c7=np.arange(0,20)
    norm7 = cm.colors.Normalize(vmin=c7.min(), vmax=c7.max())
    cmap7 = cm.ScalarMappable(norm=norm7, cmap=cm.terrain)
    cmap7.set_array([])
    
    fig2c, ax2c = plt.subplots(figsize=(10, 7))
#     plt.ticklabel_format(axis="x", style="sci", scilimits=(0,0))
    plt.suptitle("Pedersen Conductivity")
    ax2c.set_title('%s Lattitude=%s Longitude=%s' % (time_plot.strftime("%d %b %Y %H:%M:%S"),lat_out[0],lon_out[0]))
#     ax1b.set_xscale('log')
    ax2c.plot(sigmaPed1[0:-1],zg1[0:-1],c='black', linewidth=2, label='Pedersen')
    ax2c.plot(sigmaPed_Op1[0:-1],zg1[0:-1],c='tab:blue', linewidth=2, label='O+')
    ax2c.plot(sigmaPed_O2p1[0:-1],zg1[0:-1],c='tab:red', linewidth=2, label='O2+')
    ax2c.plot(sigmaPed_NOp1[0:-1],zg1[0:-1],c='tab:green', linewidth=2, label='NO+')
    ax2c.plot(sigmaPed_e1[0:-1],zg1[0:-1],c='fuchsia', linewidth=2, label='e-')
    ax2c.plot(sigmaPed_O2p_Ο21[0:-1],zg1[0:-1],c='orangered', linewidth=2, label='O2+_O2')
    ax2c.plot(sigmaPed_O2p_Ο1[0:-1],zg1[0:-1],c='lightcoral', linewidth=2, label='O2+_O')
    ax2c.plot(sigmaPed_O2p_N21[0:-1],zg1[0:-1],c='lightsalmon', linewidth=2, label='O2+_N2')
    ax2c.plot(sigmaPed_Op_Ο21[0:-1],zg1[0:-1],c='tab:cyan', linewidth=2, label='O+_O2')
    
    ax2c.plot(sigmaPed_Op_Ο1[0:-1],zg1[0:-1],c='cornflowerblue', linewidth=2, label='O+_O')
    ax2c.plot(sigmaPed_Op_N21[0:-1],zg1[0:-1],c='dodgerblue', linewidth=2, label='O+_N2')
    ax2c.plot(sigmaPed_NOp_Ο21[0:-1],zg1[0:-1],c='limegreen', linewidth=2, label='NO+_O2')
    ax2c.plot(sigmaPed_NOp_Ο1[0:-1],zg1[0:-1],c='springgreen', linewidth=2, label='NO+_O')
    ax2c.plot(sigmaPed_NOp_N21[0:-1],zg1[0:-1],c='lawngreen', linewidth=2, label='NO+_N2')
    ax2c.grid(True, color="#93a1a1", alpha=0.3)
    ax2c.minorticks_on()
    plt.legend(loc='best')
    plt.ylim(100,400)
    ax2c.set_xlabel("Conductivity (S/m)", labelpad=15, fontsize=12, color="#333533")
    ax2c.set_ylabel("Altitude (km)", labelpad=15, fontsize=12, color="#333533")

    plt.show()
######################
    
    fig2c, ax2c = plt.subplots(figsize=(10, 7))
#     plt.ticklabel_format(axis="x", style="sci", scilimits=(0,0))
    plt.suptitle("Pedersen Conductivity")
    ax2c.set_title('%s Lattitude=%s Longitude=%s' % (time_plot.strftime("%d %b %Y %H:%M:%S"),lat_out[0],lon_out[0]))
#     ax1b.set_xscale('log')
    ax2c.plot(sigmaPed_f08_out[0:-1],zg1[0:-1],c='black', linewidth=2, label='fcor=0.8')
    ax2c.plot(sigmaPed_f09_out[0:-1],zg1[0:-1],c='tab:blue', linewidth=2, label='fcor=0.9')
    ax2c.plot(sigmaPed_f10_out[0:-1],zg1[0:-1],c='tab:red', linewidth=2, label='fcor=1.0')
    ax2c.plot(sigmaPed_f11_out[0:-1],zg1[0:-1],c='tab:green', linewidth=2, label='fcor=1.1')
    ax2c.plot(sigmaPed_f12_out[0:-1],zg1[0:-1],c='fuchsia', linewidth=2, label='fcor=1.2')
    ax2c.plot(sigmaPed_f13_out[0:-1],zg1[0:-1],c='orangered', linewidth=2, label='fcor=1.3')
    ax2c.plot(sigmaPed_f14_out[0:-1],zg1[0:-1],c='lightcoral', linewidth=2, label='fcor=1.4')
    ax2c.plot(sigmaPed_f15_out[0:-1],zg1[0:-1],c='lightsalmon', linewidth=2, label='fcor=1.5')
    ax2c.plot(sigmaPed_f16_out[0:-1],zg1[0:-1],c='tab:cyan', linewidth=2, label='fcor=1.6')
    
    ax2c.plot(sigmaPed_f17_out[0:-1],zg1[0:-1],c='cornflowerblue', linewidth=2, label='fcor=1.7')
    ax2c.plot(sigmaPed_f18_out[0:-1],zg1[0:-1],c='dodgerblue', linewidth=2, label='fcor=1.8')
    ax2c.grid(True, color="#93a1a1", alpha=0.3)
    ax2c.minorticks_on()
    plt.legend(loc='best')
    plt.ylim(100,400)
    ax2c.set_xlabel("Conductivity (S/m)", labelpad=15, fontsize=12, color="#333533")
    ax2c.set_ylabel("Altitude (km)", labelpad=15, fontsize=12, color="#333533")

    plt.show()

    fig2c, ax2c = plt.subplots(figsize=(10, 7))
#     plt.ticklabel_format(axis="x", style="sci", scilimits=(0,0))
    plt.suptitle("Hall Conductivity")
    ax2c.set_title('%s Lattitude=%s Longitude=%s' % (time_plot.strftime("%d %b %Y %H:%M:%S"),lat_out[0],lon_out[0]))
#     ax1b.set_xscale('log')
    ax2c.plot(sigmaHall_f08_out[0:-1],zg1[0:-1],c='black', linewidth=2, label='fcor=0.8')
    ax2c.plot(sigmaHall_f09_out[0:-1],zg1[0:-1],c='tab:blue', linewidth=2, label='fcor=0.9')
    ax2c.plot(sigmaHall_f10_out[0:-1],zg1[0:-1],c='tab:red', linewidth=2, label='fcor=1.0')
    ax2c.plot(sigmaHall_f11_out[0:-1],zg1[0:-1],c='tab:green', linewidth=2, label='fcor=1.1')
    ax2c.plot(sigmaHall_f12_out[0:-1],zg1[0:-1],c='fuchsia', linewidth=2, label='fcor=1.2')
    ax2c.plot(sigmaHall_f13_out[0:-1],zg1[0:-1],c='orangered', linewidth=2, label='fcor=1.3')
    ax2c.plot(sigmaHall_f14_out[0:-1],zg1[0:-1],c='lightcoral', linewidth=2, label='fcor=1.4')
    ax2c.plot(sigmaHall_f15_out[0:-1],zg1[0:-1],c='lightsalmon', linewidth=2, label='fcor=1.5')
    ax2c.plot(sigmaHall_f16_out[0:-1],zg1[0:-1],c='tab:cyan', linewidth=2, label='fcor=1.6')
    
    ax2c.plot(sigmaHall_f17_out[0:-1],zg1[0:-1],c='cornflowerblue', linewidth=2, label='fcor=1.7')
    ax2c.plot(sigmaHall_f18_out[0:-1],zg1[0:-1],c='dodgerblue', linewidth=2, label='fcor=1.8')
    ax2c.grid(True, color="#93a1a1", alpha=0.3)
    ax2c.minorticks_on()
    plt.legend(loc='best')
    plt.ylim(100,400)
    plt.xlim(0,0.000002)
    ax2c.set_xlabel("Conductivity (S/m)", labelpad=15, fontsize=12, color="#333533")
    ax2c.set_ylabel("Altitude (km)", labelpad=15, fontsize=12, color="#333533")

    plt.show()
    
#####################
    fig2c, ax2c = plt.subplots(figsize=(10, 7))
#     plt.ticklabel_format(axis="x", style="sci", scilimits=(0,0))
    plt.suptitle("Hall Conductivity")
    ax2c.set_title('%s Lattitude=%s Longitude=%s' % (time_plot.strftime("%d %b %Y %H:%M:%S"),lat_out[0],lon_out[0]))
#     ax1b.set_xscale('log')
    ax2c.plot(sigmaHall1[0:-1],zg1[0:-1],c='black', linewidth=2, label='Hall')
#     ax2c.plot(sigmaHall_Op1[0:-1],zg1[0:-1],c='tab:blue', linewidth=2, label='O+')
#     ax2c.plot(sigmaHall_O2p1[0:-1],zg1[0:-1],c='tab:red', linewidth=2, label='O2+')
#     ax2c.plot(sigmaHall_NOp1[0:-1],zg1[0:-1],c='tab:green', linewidth=2, label='NO+')
#     ax2c.plot(sigmaHall_e1[0:-1],zg1[0:-1],c='fuchsia', linewidth=2, label='e-')
    ax2c.plot(sigmaHall_O2p_Ο21[0:-1],zg1[0:-1],c='orangered', linewidth=2, label='O2+_O2')
    ax2c.plot(sigmaHall_O2p_Ο1[0:-1],zg1[0:-1],c='lightcoral', linewidth=2, label='O2+_O')
    ax2c.plot(sigmaHall_O2p_N21[0:-1],zg1[0:-1],c='lightsalmon', linewidth=2, label='O2+_N2')
    ax2c.plot(sigmaHall_Op_Ο21[0:-1],zg1[0:-1],c='tab:cyan', linewidth=2, label='O+_O2')
#     
    ax2c.plot(sigmaHall_Op_Ο1[0:-1],zg1[0:-1],c='cornflowerblue', linewidth=2, label='O+_O')
    ax2c.plot(sigmaHall_Op_N21[0:-1],zg1[0:-1],c='dodgerblue', linewidth=2, label='O+_N2')
    ax2c.plot(sigmaHall_NOp_Ο21[0:-1],zg1[0:-1],c='limegreen', linewidth=2, label='NO+_O2')
    ax2c.plot(sigmaHall_NOp_Ο1[0:-1],zg1[0:-1],c='springgreen', linewidth=2, label='NO+_O')
    ax2c.plot(sigmaHall_NOp_N21[0:-1],zg1[0:-1],c='lawngreen', linewidth=2, label='NO+_N2')
    ax2c.grid(True, color="#93a1a1", alpha=0.3)
    ax2c.minorticks_on()
    plt.legend(loc='best')
    plt.ylim(100,400)
    ax2c.set_xlabel("Conductivity (S/m)", labelpad=15, fontsize=12, color="#333533")
    ax2c.set_ylabel("Altitude (km)", labelpad=15, fontsize=12, color="#333533")

    plt.show()


    fig1c, ax1c = plt.subplots(figsize=(10, 7))
#     plt.ticklabel_format(axis="x", style="sci", scilimits=(0,0))
    ax1c.set_title(r'$O_+-O$ Collision Frequencies')
#     ax1b.set_xscale('log')
    ax1c.plot(Tr_out[0:-1],q_dalgarno_out[0:-1],color='tab:blue', linewidth=2, label='Dalgarno_1964')
    ax1c.plot(Tr_out[0:-1],q_banks_out[0:-1],color='tab:orange', linewidth=2, label='Banks_1966')
    ax1c.plot(Tr_out[0:-1],q_stubbe_out[0:-1],color='tab:red', linewidth=2, label='Stubbe_1968')
    ax1c.plot(Tr_out[0:-1],q_sw_out[0:-1],color='tab:green', linewidth=2, label='Shcunk & Walker_1973')
    ax1c.plot(Tr_out[0:-1],q_salah_out[0:-1],color='tab:purple', linewidth=2, label='Salah_1993')
    ax1c.plot(Tr_out[0:-1],q_pesnell_out[0:-1],color='tab:pink', linewidth=2, label='Pesnell_1993')
    ax1c.plot(Tr_out[0:-1],q_hickmann_out[0:-1],color='tab:olive', linewidth=2, label='Hickman_1997')
    ax1c.plot(Tr_out[0:-1],q_shunk_nagy_out[0:-1],color='tab:gray', linewidth=2, label='Schunk $ Nagy_2009')
    ax1c.plot(Tr_out[0:-1],q_richmond_out[0:-1],color='tab:cyan', linewidth=2, label='Richmond_2006')
    ax1c.grid(True, color="#93a1a1", alpha=0.3)
    ax1c.minorticks_on()
    plt.legend(loc='best')
#     plt.ylim(100,400)
    ax1c.set_xlabel("Tr (K)", labelpad=15, fontsize=12, color="#333533")
    ax1c.set_ylabel(r"$\nu_{O+-O}/N_O  (cm^{-3} s^{-1}$)", labelpad=15, fontsize=12, color="#333533")

    plt.show()

    fig1c, ax1c = plt.subplots(figsize=(10, 7))
#     plt.ticklabel_format(axis="x", style="sci", scilimits=(0,0))
    ax1c.set_title(r'$O_+-O$ Collision Frequencies')
#     ax1b.set_xscale('log')
    ax1c.plot(Tr_iri[0:-1],q_dalgarno_iri[0:-1],color='tab:blue', linewidth=2, label='Dalgarno_1964')
    ax1c.plot(Tr_iri[0:-1],q_banks_iri[0:-1],color='tab:orange', linewidth=2, label='Banks_1966')
    ax1c.plot(Tr_iri[0:-1],q_stubbe_iri[0:-1],color='tab:red', linewidth=2, label='Stubbe_1968')
    ax1c.plot(Tr_iri[0:-1],q_sw_iri[0:-1],color='tab:green', linewidth=2, label='Shcunk & Walker_1973')
    ax1c.plot(Tr_iri[0:-1],q_salah_iri[0:-1],color='tab:purple', linewidth=2, label='Salah_1993')
    ax1c.plot(Tr_iri[0:-1],q_pesnell_iri[0:-1],color='tab:pink', linewidth=2, label='Pesnell_1993')
    ax1c.plot(Tr_iri[0:-1],q_hickmann_iri[0:-1],color='tab:olive', linewidth=2, label='Hickman_1997')
    ax1c.plot(Tr_iri[0:-1],q_shunk_nagy_iri[0:-1],color='tab:gray', linewidth=2, label='Schunk $ Nagy_2009')
    ax1c.plot(Tr_iri[0:-1],q_richmond_iri[0:-1],color='tab:cyan', linewidth=2, label='Richmond_2006')
    ax1c.grid(True, color="#93a1a1", alpha=0.3)
    ax1c.minorticks_on()
    plt.legend(loc='best')
#     plt.ylim(100,400)
    ax1c.set_xlabel("Tr (K)", labelpad=15, fontsize=12, color="#333533")
    ax1c.set_ylabel(r"$\nu_{O+-O}/N_O  (cm^{-3} s^{-1}$)", labelpad=15, fontsize=12, color="#333533")

    plt.show()

    fig1c, ax1c = plt.subplots(figsize=(10, 7))
#     plt.ticklabel_format(axis="x", style="sci", scilimits=(0,0))
    ax1c.set_title(r'$O_+-O$ Collision Frequencies')
#     ax1b.set_xscale('log')
    ax1c.plot(q_dalgarno_iri[0:-1],zg1[0:-1],color='tab:blue', linewidth=2, label='Dalgarno_1964')
    ax1c.plot(q_banks_iri[0:-1],zg1[0:-1],color='tab:orange', linewidth=2, label='Banks_1966')
    ax1c.plot(q_stubbe_iri[0:-1],zg1[0:-1],color='tab:red', linewidth=2, label='Stubbe_1968')
    ax1c.plot(q_sw_iri[0:-1],zg1[0:-1],color='tab:green', linewidth=2, label='Shcunk & Walker_1973')
    ax1c.plot(q_salah_iri[0:-1],zg1[0:-1],color='tab:purple', linewidth=2, label='Salah_1993')
    ax1c.plot(q_pesnell_iri[0:-1],zg1[0:-1],color='tab:pink', linewidth=2, label='Pesnell_1993')
    ax1c.plot(q_hickmann_iri[0:-1],zg1[0:-1],color='tab:olive', linewidth=2, label='Hickman_1997')
    ax1c.plot(q_shunk_nagy_iri[0:-1],zg1[0:-1],color='tab:gray', linewidth=2, label='Schunk $ Nagy_2009')
    ax1c.plot(q_richmond_iri[0:-1],zg1[0:-1],color='tab:cyan', linewidth=2, label='Richmond_2006')
    ax1c.grid(True, color="#93a1a1", alpha=0.3)
    ax1c.minorticks_on()
    plt.legend(loc='best')
#     plt.ylim(100,400)
    ax1c.set_xlabel(r"$\nu_{O+-O}/N_O  (cm^{-3} s^{-1}$)", labelpad=15, fontsize=12, color="#333533")
    ax1c.set_ylabel("Altitude (km)", labelpad=15, fontsize=12, color="#333533")

    plt.show()
    
def plot_frequencies(timer):

    time_begin=datetime.datetime(2015,3,15,0,0,0)
    time_plot=time_begin+datetime.timedelta(minutes=maptime[timer])
    
    fig1 = go.Figure()
    fig1.add_trace(go.Scatter(x=omega_e1[0:-1], y=zg1[0:-1],name="$$\Omega_{e}$$",mode='lines',marker=dict(color='blue')))
    fig1.add_trace(go.Scatter(x=omega_NOp1[0:-1], y=zg1[0:-1],name="$$\Omega_{NO+}$$",mode='lines',marker=dict(color='crimson')))
    fig1.add_trace(go.Scatter(x=omega_Op1[0:-1], y=zg1[0:-1],name="$$\Omega_{O+}$$",mode='lines',marker=dict(color='green')))
    fig1.add_trace(go.Scatter(x=omega_O2p1[0:-1], y=zg1[0:-1],name="$$\Omega_{O2+}$$+",mode='lines',marker=dict(color='orange')))
    fig1.add_trace(go.Scatter(x=vOp1[0:-1], y=zg1[0:-1],name="$$v_{O+}$$",mode='lines',marker=dict(color='mediumseagreen')))
    fig1.add_trace(go.Scatter(x=ven1[0:-1], y=zg1[0:-1],name="$$v_{en}$$",mode='lines',marker=dict(color='dodgerblue')))
    fig1.add_trace(go.Scatter(x=vO2p1[0:-1], y=zg1[0:-1],name="$$v_{O2+}$$+",mode='lines',marker=dict(color='sandybrown')))
    fig1.add_trace(go.Scatter(x=vNOp1[0:-1], y=zg1[0:-1],name="$$v_{NO+}$$+",mode='lines',marker=dict(color='deeppink')))
    fig1.add_trace(go.Scatter(x=vin1[0:-1], y=zg1[0:-1],name="$$v_{in}$$+",mode='lines',marker=dict(color='mediumpurple')))
    fig1.add_trace(go.Scatter(x=vei[0:-1], y=zg1[0:-1],name="$$v_{ei}$$+",mode='lines',marker=dict(color='gold')))
    fig1.add_trace(go.Scatter(x=vop_n_SB[0:-1], y=zg1[0:-1],name="$$v_{O+} SB$$+",mode='lines',marker=dict(color='magenta')))
    fig1.add_trace(go.Scatter(x=vo2p_n_SB[0:-1], y=zg1[0:-1],name="$$v_{O2+} SB$$+",mode='lines',marker=dict(color='brown')))
    
    fig1.update_layout(xaxis_type="log",xaxis_showexponent='all', xaxis_exponentformat = 'power',
    yaxis_tick0 = 60,yaxis_dtick = 20,xaxis_title="Frequency (Hz)",yaxis_title="Altitude (km)",width=750,height=750,
    title={'text':'Frequencies %s <br> lattitude=%s longitude=%s' % (time_plot.strftime("%d %b %Y %H:%M:%S"),lat_out[0],lon_out[0]),
           'y':0.9,'x':0.5,'xanchor': 'center','yanchor': 'top'})    
    fig1.update_xaxes(showgrid=True, gridwidth=0.5, gridcolor='grey')
    fig1.update_yaxes(showgrid=True, gridwidth=0.5, gridcolor='grey')
    fig1.update_xaxes(showline=True, linewidth=2, linecolor='black', mirror=True)
    fig1.update_yaxes(showline=True, linewidth=2, linecolor='black', mirror=True)
    fig1.show()   
    
    
 
    
def Exec_Btn_Clicked( b ):
    print( "Calculation started..." )
    run(tiegcm_file.value,lat_value.value,lon_value.value,timer_value.value)
    if conductivities_checkbox.value == True:
        print( "Plotting conductivities..." )
        collisions()
        plot_conductivities(timer_value.value)
        
    if frequencies_checkbox.value == True:
        print( "Plotting frequencies..." )
        collisions()
        plot_frequencies(timer_value.value)
        
        
  
    
    
def createGUI():
    ## the top level visual elements
    MainPanel = widgets.VBox() 
    MainTab = widgets.Tab() 
    VerticalPanel = widgets.VBox()
    MapPanel = widgets.VBox()
    LatAltPanel= widgets.VBox()
    
    MainTab.children = [ VerticalPanel]
    MainTab.set_title(0, 'Vertical Profiles')
    MainTab.set_title(1, 'Maps')
    MainTab.set_title(2, 'Latitude-Altitude')
    MainPanel.children = [MainTab ]   
    

    PlotVerticalPanel = widgets.VBox()
    PlotVerticalPanel.children = [lat_value,lon_value,timer_value,widgets.VBox([conductivities_checkbox,frequencies_checkbox])]
    
   
    Exec_Btn = widgets.Button (description='Calculate Products',tooltip="Click here to calculate Daedalus products",)
    Exec_Btn.style.button_color = 'MediumTurquoise'
    Exec_Btn.on_click( Exec_Btn_Clicked )
    VerticalPanel.children = [tiegcm_file, PlotVerticalPanel, Exec_Btn ]
    

    return MainPanel
display( createGUI() )
```
