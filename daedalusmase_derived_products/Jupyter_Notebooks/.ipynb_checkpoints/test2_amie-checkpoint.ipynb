```python
#import required packages
import sys
from netCDF4 import Dataset
import numpy as np
import datetime
import pyglow
from ipywidgets import*
import ipywidgets as widgets
import warnings
import glob
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from mpl_toolkits.basemap import Basemap, shiftgrid, addcyclic
import matplotlib.mlab as mlab
from mpl_toolkits.axes_grid1 import make_axes_locatable
from scipy.interpolate import griddata
from pylab import *
from matplotlib.patches import Polygon
import time
import pandas as pd
from scipy.special import erf
import matplotlib.colors as colors
from matplotlib import ticker, cm

#Define WPIT package location
current_dir =  os.path.abspath(os.path.dirname('__file__'))
fpath = os.path.abspath(current_dir + "/..")
sys.path.append(fpath)
sys.path.append(current_dir)

#import sub_Collision_Freqs_Cross_Sections sub-module
import DaedalusMASE_Derived_Products.sub_Collision_Freqs_Cross_Sections as cols

#import sub_Conductivities sub-module
import DaedalusMASE_Derived_Products.sub_Conductivities as cond

#import sub_coords_utils sub-module
import DaedalusMASE_Derived_Products.sub_coords_utils as coords

#import sub_Currents_Magnetic_Forcing sub-module
import DaedalusMASE_Derived_Products.sub_Currents_Magnetic_Forcing as currents

#import sub_Heating_Sources sub-module
import DaedalusMASE_Derived_Products.sub_Heating_Sources as heatings


#import sub_Heating_Sources sub-module
import DaedalusMASE_Derived_Products.sub_tiegcm_utils as tiegcm


#Allocations
Ohmic1=[]
zg1=[]
Joule1=[]
Frictional1=[]
nO1=[]
nO21=[]
nN21=[]
Ne1=[]
NOp1=[]
NO2p1=[]
sigmaHall1=[]
sigmaPed1=[]
sigmaHall1TIEGCM=[]
sigmaPed1TIEGCM=[]
sigma01=[]
omega_e1=[]
omega_Op1=[]
vOp1=[]
ven1=[]
omega_O2p1=[]
JH_rhov=[]
vNOp1=[]
NNOp1=[]
omega_NOp1=[]
vO2p1=[]
vin1=[]
vi_op_x1=[]
vi_op_y1=[]
vi_op_z1=[]
vi_o2p_x1=[]
vi_o2p_y1=[]
vi_o2p_z1=[]
vi_nop_x1=[]
vi_nop_y1=[]
vi_nop_z1=[]
unvx1=[]
unvy1=[]
unvz1=[]
ExBx1=[]
ExBy1=[]
ExBz1=[]
ve_x1=[]
ve_y1=[]
ve_z1=[]
E_e=[]
E_n=[]
E_u=[]
J_pe=[]
J_pn=[]
J_pu=[]
J_he=[]
J_hn=[]
J_hu=[]
T_e=[]
T_i=[]
T_n=[]
J_p_Ohmice=[]
J_p_Ohmicn=[]
J_p_Ohmicu=[]
J_dene=[]
J_denn=[]
J_denu=[]
cross_sectionf=[]
Pressure=[]
ohmic_pressure=[]
pressure_ohmic=[]
lat_out=[]
lon_out=[]
wind1=[]
convection1=[]
ohms_ohmic=[]
Ee_TIEGCM1=[]
En_TIEGCM1=[]
Eu_TIEGCM1=[]
QAMIE1=[]
QWIND1=[]
cross_op_sectionf=[]
cross_nop_sectionf=[]
cross_o2p_sectionf=[]
qDTop_o=[]
qDTop_o2=[]
qDTop_n2=[]
qDTop_he=[]
qDTo2p_o=[]
qDTo2p_o2=[]
qDTo2p_n2=[]
qDTo2p_he=[]
qDTnop_o=[]
qDTnop_o2=[]
qDTnop_n2=[]
qDTnop_he=[]
qDTop_n=[]
qDTo2p_n=[]
qDTnop_n=[]
qDTi_n=[]
Argon=[]
Helium=[]
NitricOxide=[]
Nitrogen_ion=[]
Le_N2_rot=[]
Le_O2_rot=[]
Le_O_fine=[]
Le_N2_vib=[]
Le_O_1D=[]
Le_O2_vib=[]
qDTe_o=[]
qDTe_o2=[]
qDTe_n2=[]
qDTe_he=[]
qDTe_n=[]
Le1=[]
vei=[]
qDTe_op=[]
qDTe_o2p=[]
qDTe_np=[]
qDTe_nop=[]
qDTe_i=[]
vop_n_SB=[]
vo2p_n_SB=[]

Nef=np.zeros((72, 144),order='F')
NOf=np.zeros((72, 144),order='F')
NO2f=np.zeros((72, 144),order='F')
NN2f=np.zeros((72, 144),order='F')
NNOpf=np.zeros((72, 144),order='F')
Evert=np.zeros((72, 144),order='F')
Evertx=np.zeros((72, 144),order='F')
Everty=np.zeros((72, 144),order='F')
Evertz=np.zeros((72, 144),order='F')
NOpf=np.zeros((72, 144),order='F')
NO2pf=np.zeros((72, 144),order='F')
Omega_ef=np.zeros((72, 144),order='F')
Omega_Opf=np.zeros((72, 144),order='F')
vOpf=np.zeros((72, 144),order='F')
venf=np.zeros((72, 144),order='F')
sigmaPedf=np.zeros((72, 144),order='F')
sigmaPedf_T=np.zeros((72, 144),order='F')
sigmaHallf=np.zeros((72, 144),order='F')
sigmaHallf_T=np.zeros((72, 144),order='F')
TIEGCM_PEDf=np.zeros((72, 144),order='F')
TIEGCM_HALLf=np.zeros((72, 144),order='F')
Potf=np.zeros((72, 144),order='F')
vi_starxf_op=np.zeros((72, 144),order='F')
vi_staryf_op=np.zeros((72, 144),order='F')
vi_starzf_op=np.zeros((72, 144),order='F')
vi_starxf_o2p=np.zeros((72, 144),order='F')
vi_staryf_o2p=np.zeros((72, 144),order='F')
vi_starzf_o2p=np.zeros((72, 144),order='F')
vi_starxf_nop=np.zeros((72, 144),order='F')
vi_staryf_nop=np.zeros((72, 144),order='F')
vi_starzf_nop=np.zeros((72, 144),order='F')
Bxf=np.zeros((72, 144),order='F')
Byf=np.zeros((72, 144),order='F')
Bzf=np.zeros((72, 144),order='F')
zgf=np.zeros((72, 144),order='F')
Ohmic=np.zeros((72, 144),order='F')
Joule=np.zeros((72, 144),order='F')
Frictional=np.zeros((72, 144),order='F')
Ohmic_TIEGCM=np.zeros((72, 144),order='F')
unvertxf=np.zeros((72, 144),order='F')
unvertyf=np.zeros((72, 144),order='F')
unvertzf=np.zeros((72, 144),order='F')
ExB_x=np.zeros((72, 144),order='F')
ExB_y=np.zeros((72, 144),order='F')
ExB_z=np.zeros((72, 144),order='F')
ExBx_calc=np.zeros((72, 144),order='F')
ExBy_calc=np.zeros((72, 144),order='F')
ExBz_calc=np.zeros((72, 144),order='F')
NO2pf=np.zeros((72, 144),order='F')
Omega_NOpf=np.zeros((72, 144),order='F')
Omega_O2pf=np.zeros((72, 144),order='F')
vO2pf=np.zeros((72, 144),order='F')
vinf=np.zeros((72, 144),order='F')
JH_rho=np.zeros((72, 144),order='F')
sigma_par=np.zeros((72, 144),order='F')
J_densx=np.zeros((72, 144),order='F')
J_densy=np.zeros((72, 144),order='F')
J_densz=np.zeros((72, 144),order='F')
J_Ohmx=np.zeros((72, 144),order='F')
J_Ohmy=np.zeros((72, 144),order='F')
J_Ohmz=np.zeros((72, 144),order='F')
vee_out=np.zeros((72, 144),order='F')
ven_out=np.zeros((72, 144),order='F')
veu_out=np.zeros((72, 144),order='F')
maplat=np.zeros((72),order='F')
maplon=np.zeros((144),order='F')
maptime=np.zeros((24),order='F')
J_Pede=np.zeros((72, 144),order='F')
J_Pedn=np.zeros((72, 144),order='F')
J_Pedu=np.zeros((72, 144),order='F')
J_Halle=np.zeros((72, 144),order='F')
J_Halln=np.zeros((72, 144),order='F')
J_Hallu=np.zeros((72, 144),order='F')
J_Ohme=np.zeros((72, 144),order='F')
J_Ohmn=np.zeros((72, 144),order='F')
J_Ohmu=np.zeros((72, 144),order='F')
vi_mag_op=np.zeros((72, 144),order='F')
vi_mag_o2p=np.zeros((72, 144),order='F')
vi_mag_nop=np.zeros((72, 144),order='F')
J_mag=np.zeros((72, 144),order='F')
Unef=np.zeros((72, 144),order='F')
Unnf=np.zeros((72, 144),order='F')
Unuf=np.zeros((72, 144),order='F')
JxBe=np.zeros((72, 144),order='F')
JxBn=np.zeros((72, 144),order='F')
JxBu=np.zeros((72, 144),order='F')
QAMIE_T=np.zeros((72, 144),order='F')
QWIND_T=np.zeros((72, 144),order='F')
Ti_Out=np.zeros((72, 144),order='F')
Te_Out=np.zeros((72, 144),order='F')
Tn_Out=np.zeros((72, 144),order='F')
Press_Ohmic=np.zeros((72, 144),order='F')
cross_section=np.zeros((72, 144),order='F')
cross_section_op=np.zeros((72, 144),order='F')
cross_section_o2p=np.zeros((72, 144),order='F')
cross_section_nop=np.zeros((72, 144),order='F')
Convenction_heating_Out=np.zeros((72, 144),order='F')
Wind_heating_Out=np.zeros((72, 144),order='F')
Ohmic_heating_OhmsLaw_Out=np.zeros((72, 144),order='F')
Mechanical_power_Out=np.zeros((72, 144),order='F')
Energy_Out=np.zeros((72, 144),order='F')
Pressure_2D=np.zeros((72, 144),order='F')
Ohmic_Press=np.zeros((72, 144),order='F')
vNOpf=np.zeros((72, 144),order='F')
qDTop_of=np.zeros((72, 144),order='F')
qDTop_o2f=np.zeros((72, 144),order='F')
qDTop_n2f=np.zeros((72, 144),order='F')
qDTo2p_of=np.zeros((72, 144),order='F')
qDTo2p_o2f=np.zeros((72, 144),order='F')
qDTo2p_n2f=np.zeros((72, 144),order='F')
qDTnop_of=np.zeros((72, 144),order='F')
qDTnop_o2f=np.zeros((72, 144),order='F')
qDTnop_n2f=np.zeros((72, 144),order='F')
qDTop_nf=np.zeros((72, 144),order='F')
qDTo2p_nf=np.zeros((72, 144),order='F')
qDTnop_nf=np.zeros((72, 144),order='F')
qDTi_nf=np.zeros((72, 144),order='F')
Le_N2_rot_2d=np.zeros((72, 144),order='F')
Le_O2_rot_2d=np.zeros((72, 144),order='F')
Le_O2_vib_2d=np.zeros((72, 144),order='F')
Le_O_fine_2d=np.zeros((72, 144),order='F')
Le_N2_vib_2d=np.zeros((72, 144),order='F')
Le_O_1D_2d=np.zeros((72, 144),order='F')
Le_all_2d=np.zeros((72, 144),order='F')
qDTe_o_2d=np.zeros((72, 144),order='F')
qDTe_o2_2d=np.zeros((72, 144),order='F')
qDTe_n2_2d=np.zeros((72, 144),order='F')
qDTe_he_2d=np.zeros((72, 144),order='F')
qDTe_op_2d=np.zeros((72, 144),order='F')
qDTe_o2p_2d=np.zeros((72, 144),order='F')
qDTe_nop_2d=np.zeros((72, 144),order='F')
qDTe_np_2d=np.zeros((72, 144),order='F')
qDTe_i_2d=np.zeros((72, 144),order='F')
qDTe_n_2d=np.zeros((72, 144),order='F')

Nefl=np.zeros((57, 72),order='F')
NOfl=np.zeros((57, 72),order='F')
NO2fl=np.zeros((57, 72),order='F')
NN2fl=np.zeros((57, 72),order='F')
NNOpfl=np.zeros((57, 72),order='F')
Evertl=np.zeros((57, 72),order='F')
Evertxl=np.zeros((57, 72),order='F')
Evertyl=np.zeros((57, 72),order='F')
Evertzl=np.zeros((57, 72),order='F')
NOpfl=np.zeros((57, 72),order='F')
NO2pfl=np.zeros((57, 72),order='F')
Omega_efl=np.zeros((57, 72),order='F')
Omega_Opfl=np.zeros((57, 72),order='F')
vOpfl=np.zeros((57, 72),order='F')
venfl=np.zeros((57, 72),order='F')
sigmaPedfl=np.zeros((57, 72),order='F')
sigmaPedf_Tl=np.zeros((57, 72),order='F')
sigmaHallfl=np.zeros((57, 72),order='F')
sigmaHallf_Tl=np.zeros((57, 72),order='F')
TIEGCM_PEDfl=np.zeros((57, 72),order='F')
TIEGCM_HALLfl=np.zeros((57, 72),order='F')
Potfl=np.zeros((57, 72),order='F')
vi_starxf_opl=np.zeros((57, 72),order='F')
vi_staryf_opl=np.zeros((57, 72),order='F')
vi_starzf_opl=np.zeros((57, 72),order='F')
vi_starxf_o2pl=np.zeros((57, 72),order='F')
vi_staryf_o2pl=np.zeros((57, 72),order='F')
vi_starzf_o2pl=np.zeros((57, 72),order='F')
vi_starxf_nopl=np.zeros((57, 72),order='F')
vi_staryf_nopl=np.zeros((57, 72),order='F')
vi_starzf_nopl=np.zeros((57, 72),order='F')
Bxfl=np.zeros((57, 72),order='F')
Byfl=np.zeros((57, 72),order='F')
Bzfl=np.zeros((57, 72),order='F')
zgfl=np.zeros((57, 72),order='F')
Ohmicl=np.zeros((57, 72),order='F')
Joulel=np.zeros((57, 72),order='F')
Frictionall=np.zeros((57, 72),order='F')
Ohmic_TIEGCMl=np.zeros((57, 72),order='F')
unvertxfl=np.zeros((57, 72),order='F')
unvertyfl=np.zeros((57, 72),order='F')
unvertzfl=np.zeros((57, 72),order='F')
ExB_xl=np.zeros((57, 72),order='F')
ExB_yl=np.zeros((57, 72),order='F')
ExB_zl=np.zeros((57, 72),order='F')
ExBx_calcl=np.zeros((57, 72),order='F')
ExBy_calcl=np.zeros((57, 72),order='F')
ExBz_calcl=np.zeros((57, 72),order='F')
NO2pfl=np.zeros((57, 72),order='F')
Omega_NOpfl=np.zeros((57, 72),order='F')
Omega_O2pfl=np.zeros((57, 72),order='F')
vO2pfl=np.zeros((57, 72),order='F')
vinfl=np.zeros((57, 72),order='F')
JH_rhol=np.zeros((57, 72),order='F')
sigma_parl=np.zeros((57, 72),order='F')
J_densxl=np.zeros((57, 72),order='F')
J_densyl=np.zeros((57, 72),order='F')
J_denszl=np.zeros((57, 72),order='F')
J_Ohmxl=np.zeros((57, 72),order='F')
J_Ohmyl=np.zeros((57, 72),order='F')
J_Ohmzl=np.zeros((57, 72),order='F')
vee_outl=np.zeros((57, 72),order='F')
ven_outl=np.zeros((57, 72),order='F')
veu_outl=np.zeros((57, 72),order='F')
maplatl=np.zeros((72),order='F')
# maplon1=np.zeros((144),order='F')
mapaltl=np.zeros((57),order='F')
maptimel=np.zeros((24),order='F')
J_Pedel=np.zeros((57, 72),order='F')
J_Pednl=np.zeros((57, 72),order='F')
J_Pedul=np.zeros((57, 72),order='F')
J_Hallel=np.zeros((57, 72),order='F')
J_Hallnl=np.zeros((57, 72),order='F')
J_Hallul=np.zeros((57, 72),order='F')
J_Ohmel=np.zeros((57, 72),order='F')
J_Ohmnl=np.zeros((57, 72),order='F')
J_Ohmul=np.zeros((57, 72),order='F')
vi_mag_opl=np.zeros((57, 72),order='F')
vi_mag_o2pl=np.zeros((57, 72),order='F')
vi_mag_nopl=np.zeros((57, 72),order='F')
J_magl=np.zeros((57, 72),order='F')
Unefl=np.zeros((57, 72),order='F')
Unnfl=np.zeros((57, 72),order='F')
Unufl=np.zeros((57, 72),order='F')
JxBel=np.zeros((57, 72),order='F')
JxBnl=np.zeros((57, 72),order='F')
JxBul=np.zeros((57, 72),order='F')
QAMIE_Tl=np.zeros((57, 72),order='F')
QWIND_Tl=np.zeros((57, 72),order='F')
Ti_Outl=np.zeros((57, 72),order='F')
Te_Outl=np.zeros((57, 72),order='F')
Tn_Outl=np.zeros((57, 72),order='F')
Press_Ohmicl=np.zeros((57, 72),order='F')
cross_sectionl=np.zeros((57, 72),order='F')
cross_section_opl=np.zeros((57, 72),order='F')
cross_section_o2pl=np.zeros((57, 72),order='F')
cross_section_nopl=np.zeros((57, 72),order='F')
Convenction_heating_Outl=np.zeros((57, 72),order='F')
Wind_heating_Outl=np.zeros((57, 72),order='F')
Ohmic_heating_OhmsLaw_Outl=np.zeros((57, 72),order='F')
Mechanical_power_Outl=np.zeros((57, 72),order='F')
Energy_Outl=np.zeros((57, 72),order='F')
Pressure_2Dl=np.zeros((57, 72),order='F')
Ohmic_Pressl=np.zeros((57, 72),order='F')
vNOpfl=np.zeros((57, 72),order='F')
qDTop_ofl=np.zeros((57, 72),order='F')
qDTop_o2fl=np.zeros((57, 72),order='F')
qDTop_n2fl=np.zeros((57, 72),order='F')
qDTo2p_ofl=np.zeros((57, 72),order='F')
qDTo2p_o2fl=np.zeros((57, 72),order='F')
qDTo2p_n2fl=np.zeros((57, 72),order='F')
qDTnop_ofl=np.zeros((57, 72),order='F')
qDTnop_o2fl=np.zeros((57, 72),order='F')
qDTnop_n2fl=np.zeros((57, 72),order='F')
qDTop_nfl=np.zeros((57, 72),order='F')
qDTo2p_nfl=np.zeros((57, 72),order='F')
qDTnop_nfl=np.zeros((57, 72),order='F')
qDTi_nfl=np.zeros((57, 72),order='F')

#Transformation from ENU to ECEF
def enu_ecef(lat_phi, lon_lmd, Fe, Fn, Fup):
    fac = np.pi / 180
    lat_phi = lat_phi * fac
    lon_lmd = lon_lmd * fac

    north_temp_unit = [-np.cos(lon_lmd) * np.sin(lat_phi), - np.sin(lon_lmd) * np.sin(lat_phi), np.cos(lat_phi)]
    east_temp_unit = [-np.sin(lon_lmd), np.cos(lon_lmd), 0]
    up_temp_unit = [np.cos(lon_lmd) * np.cos(lat_phi), np.sin(lon_lmd) * np.cos(lat_phi), np.sin(lat_phi)]

    Fnorth_vector = [Fn * north_temp_unit[0], Fn * north_temp_unit[1], Fn * north_temp_unit[2]]
    Feast_vector = [Fe * east_temp_unit[0], Fe * east_temp_unit[1], Fe * east_temp_unit[2]]
    Fup_vector = [Fup * up_temp_unit[0], Fup * up_temp_unit[1], Fup * up_temp_unit[2]]

    Fx = Fnorth_vector[0] + Feast_vector[0] + Fup_vector[0]
    Fy = Fnorth_vector[1] + Feast_vector[1] + Fup_vector[1]
    Fz = Fnorth_vector[2] + Feast_vector[2] + Fup_vector[2]

    return Fx, Fy, Fz

#Transformation from ECEF to ENU
def ecef_enu(lat_phi, lon_lmd, Fx, Fy, Fz):
    fac = np.pi / 180
    lat_phi = lat_phi * fac
    lon_lmd = lon_lmd * fac
    
    x_temp_unit=[-np.sin(lon_lmd), -np.cos(lon_lmd)*np.sin(lat_phi),np.cos(lon_lmd)*np.cos(lat_phi)]
    y_temp_unit=[np.cos(lon_lmd),-np.sin(lon_lmd)*np.sin(lat_phi),np.sin(lon_lmd)*np.cos(lat_phi)]
    z_temp_unit=[0, np.cos(lat_phi),np.sin(lat_phi)]

    Fx_vector = [Fx * x_temp_unit[0], Fx * x_temp_unit[1], Fx * x_temp_unit[2]]
    Fy_vector = [Fy * y_temp_unit[0], Fy * y_temp_unit[1], Fy * y_temp_unit[2]]
    Fz_vector = [Fz * z_temp_unit[0], Fz * z_temp_unit[1], Fz * z_temp_unit[2]]

    Fe = Fx_vector[0] + Fy_vector[0] + Fz_vector[0]
    Fn = Fx_vector[1] + Fy_vector[1] + Fz_vector[1]
    Fu = Fx_vector[2] + Fy_vector[2] + Fz_vector[2]

    return Fe, Fn, Fu

 # calculate geocentric latitude from geodetic latitude (WGS84)
def geo_lat2geod_lat(phi):
    a = 6378137  # meter semi major axis of earth
    f = 1 / 298.257  # flattening
    b = a - f * a  # semi minor axis
    e = ((a ** 2 - b ** 2) ** (1 / 2)) / a
    phi_rad = np.deg2rad(phi)
    geod_lat = np.arctan(np.tan(phi_rad) / (1 - e ** 2))
    geod_lat = np.rad2deg(geod_lat)
    return geod_lat  # in degrees

#Geographic to Geodetic transformation
def geo2geod(lat_geo_phi, lon_geo_lmd, alt_geo):
    lat_geod_phi = geo_lat2geod_lat(lat_geo_phi)  # degrees
    lon_geod_lmd = lon_geo_lmd  # degrees
    alt_geod = alt_geo  # km
    return lat_geod_phi, lon_geod_lmd, alt_geod

#Calculation of products for map plotting
                    
def run_map(tiegcm_file,timer_value,pressure_level):
    # Initializations
    start_time = time.time()
    time_igrf = datetime.datetime(2015, 3, 15, 0, 0, 0)  # first time step of the TIEGCM run
    timer = 0  # counter used for incrementing TIEGCM timestep
    deltalmd = 2.5  # TIEGCM grid resolution in deg
    deltaphi = 2.5  # TIEGCM grid resolution in deg
    Re = 6378.1370  # Earths Radius (km)
    electron = 1.602176565 * 10 ** (- 19) #electron charge in C
    boltzmann_si=1.38064852 * 10**(-23)
    boltzmann=1.380645852 * 10 ** (-16) #Boltzmann constant in cm^2*g*s^(-2)*K^(-1)
    me=9.10938356 * 10 ** (- 31) #electron mass in kg
    mO=16 # Oxygen atomic mass in g/mol
    mN2=28  #N2 molecular mass in g/mol
    mO2=32 #O2 molecular mass in g/mol
    mNO=30 #NO molecular mass in g/mol
    NA=6.02214086 * 10 ** 23 #Avogadro's constant in mol^-1
    fcor=1.5 #Burnside facor (default=1.5)
    warnings.simplefilter('ignore')
    #Input file
    TIEGCM = Dataset(tiegcm_file)
    
    # Get data from TIEGCM Model file
    glat = TIEGCM.variables['lat'][:] #geographic latitude in deg
    glon = TIEGCM.variables['lon'][:] #geographic longitude in deg
    glev = TIEGCM.variables['lev'][:] #interface levels
    gtime = TIEGCM.variables['time'][:] #time in minutes since 2015 March 15 00:00:00
    zg = TIEGCM.variables['ZGMID'][:] #Geometric height in cm
    Ne_all = TIEGCM.variables['NE'][:]  # electron density in cm^-3
    Ne_mid = TIEGCM.variables['ELECDEN'][:]  # electron density in cm^-3
    PED_all = TIEGCM.variables['PEDERSEN'][:] #Pedersen conductivity in Si/m
    HALL_all= TIEGCM.variables['HALL'][:] #Hall conductivity in Si/m
    Un_east = TIEGCM.variables['UN'][:]  # neutral zonal wind (+East) in cm/s
    Un_north = TIEGCM.variables['VN'][:]  # neutral meridional wind (+North) in cm/s
    Un_up = TIEGCM.variables['WN_lev'][:]  # neutral vertical wind (+Up) in cm/s
    Ui_east = TIEGCM.variables['Ui_lev'][:]  # zonal ExB velocity in cm/s
    Ui_north = TIEGCM.variables['Vi_lev'][:]  # meridional ExB velocity in cm/s
    Ui_up = TIEGCM.variables['Wi_lev'][:]  # vertical ExB velocity in cm/s
    Poten_all=TIEGCM.variables['POTEN'][:] #electric potential in V
    barm_all=TIEGCM.variables['BARM'][:] #Mean molecular weight in g/mol
    P0=TIEGCM.variables['p0'][:] #Reference pressure in millibar
    Zh=TIEGCM.variables['Z'][:] #dimensionless variable (for pressure calculations)
    Tn_all=TIEGCM.variables['TN'][:] #Neutral Temperature in K
    Te_all=TIEGCM.variables['TE'][:] #Electron Temperature in K
    Ti_all=TIEGCM.variables['TI'][:] #Ion Temperature in K
    nO_all=TIEGCM.variables['O1'][:] #neutral atomic oxygen density in mmr
    nO2_all=TIEGCM.variables['O2'][:] #neutral molecular oxygen density in mmr
    NOp_all=TIEGCM.variables['OP'][:] #O+ density in cm^-3
    den_all=TIEGCM.variables['DEN'][:] #Density
    nO_all_cm3=TIEGCM.variables['O_CM3'][:] #neutral atomic oxygen density in mmr
    nO2_all_cm3=TIEGCM.variables['O2_CM3'][:] #neutral molecular oxygen density in mmr
    NO2p_all=TIEGCM.variables['O2P'][:] #O+ density in cm^-3
    Np_all=TIEGCM.variables['NPLUS'][:] #N+ density in cm^-3
    NNOp_all=TIEGCM.variables['NOP_LAM'][:] #NO+ density in cm^-3
    N2_all_cm3=TIEGCM.variables['N2_CM3'][:] #N2 density in cm^-3
    EEX=TIEGCM.variables['EEX'][:] 
    EEY=TIEGCM.variables['EEY'][:] 
    EEZ=TIEGCM.variables['EEZ'][:] 
    QAMIE=TIEGCM.variables['QAMIE'][:] 
    QWIND=TIEGCM.variables['QWIND'][:]
    Helium_all=TIEGCM.variables['HE'][:]
    TIEGCM.close() #close input file


    for timer in range(timer_value,timer_value+1):
        time_plot=gtime[timer]
        maptime[timer]=time_plot
        
        for lev in range(pressure_level, pressure_level+1):
        
            for lat in range(0, len(glat)):

                for lon in range(0, len(glon)):
                    
                    ########################################################################
#                   Coordinates
                    ########################################################################
                    
                    # GEO coordinates of desired point
                    alt_p = zg[timer, lev, lat, lon] / 1e5 #cm to km
                    zgf[lat, lon]=alt_p #output 
                    lat_p = glat[lat]   #deg
                    lon_p = glon[lon]   #deg
                    maplat[lat]=glat[lat] #output 
                    maplon[lon]=glon[lon] #output 
                    timeg=gtime[timer] #minutes
                    #time
                    time_p = time_igrf + datetime.timedelta(minutes=timeg)               
                   
                    #Pressure levels
                    Z=glev[lev]
                    
                    pressuref=P0*np.exp(-Z) #pressure in millibars
                    pressuref_Pa=pressuref*100 #pressure in Pascals
                    Pressure_2D[lat,lon]=pressuref_Pa
                    
                    #Electric Potential (V)
                    Pot=Poten_all[timer, lev, lat, lon]
                    Potf[lat, lon]=Pot #output 
                    
                    ########################################################################
#                   Temperatures and densities
                    ########################################################################

                    #Temperatures
                    Tn=Tn_all[timer, lev, lat, lon]     #Kelvin
                    Te=Te_all[timer, lev, lat, lon]     #Kelvin
                    Ti=Ti_all[timer, lev, lat, lon]     #Kelvin
                    
                    Ti_Out[lat,lon]=Ti
                    Te_Out[lat,lon]=Te
                    Tn_Out[lat,lon]=Tn
                    
                    #mean molecular mass g/mol
                    barm=barm_all[timer, lev, lat, lon]

                    #Conversion factor (mmr->cm^(-3))
                    Nbarm=(P0*np.exp(-Z)*barm*1000)/(boltzmann*Tn)

                    #neutral Densities (cm^-3)
                    NO=nO_all_cm3[timer, lev, lat, lon] #cm^-3
                    NOf[lat, lon]=NO* (10 ** 6) #output m^-3
                    NO2=nO2_all_cm3[timer, lev, lat, lon] #cm^-3                 
                    NO2f[lat, lon]=NO2* (10 ** 6) #output m^-3
                    NN2=N2_all_cm3[timer, lev, lat, lon] #cm^-3
                    NN2f[lat, lon]=NN2* (10 ** 6) #output m^-3

                    # electron density
                    Ne = Ne_all[timer, lev, lat, lon] * (10 ** 6)  # m^-3
                    Nef[lat, lon]=Ne #output m^-3
                    Necm=Ne_all[timer, lev, lat, lon] # (cm^-3)

                    #ion Densities
                    NOp=NOp_all[timer, lev, lat, lon]  #cm^-3
                    NO2p=NO2p_all[timer, lev, lat, lon]  #cm^-3
                    NOpf[lat, lon]=NOp #output 
                    NO2pf[lat, lon]=NOp #output 
                    Nplus=Np_all[timer, lev, lat, lon]
                    NNOp=Necm-NOp+NO2p  #if O2+ density is available in the input file
                    NNOpf[lat, lon]=NNOp #output 
                    
                    rmass_n4s = 14
                    rmass_n2  = 28
                    
                    MO=mO/(NA*1000)  # Atomic oxygen mass in kg
                    MO2=(2*mO)/(NA*1000) # Molecular oxygen mass in kg
                    MNO=(mNO)/(NA*1000) #Nitric oxide mass in kg                    
                    MN2=rmass_n2/(NA*1000)   #Nitrogen mass in kg
                    MN=rmass_n4s/(NA*1000)
                    
                    
                    ########################################################################
#                   Velocities
                    ########################################################################

                    #neutral wind 
                    Une=Un_east[timer, lev, lat, lon]
                    Unn=Un_north[timer, lev, lat, lon]
                    Unu=Un_up[timer, lev, lat, lon]

                    Un_x = Une / 100  # m/s
                    Un_y = Unn / 100  # m/s
                    Un_z = Unu / 100  # m/s
                    Un = [Un_x, Un_y, Un_z] # neutral wind in ENU in m/s
                    
                    Unef[lat,lon]=Un_x #output (ENU)
                    Unnf[lat,lon]=Un_y #output (ENU)  
                    Unuf[lat,lon]=Un_z #output (ENU)  
                    

                    # ExB velocity from ENU to ECEF
                    Uie=Ui_east[timer, lev, lat, lon]
                    Uin=Ui_north[timer, lev, lat, lon]
                    Uiu=Ui_up[timer, lev, lat, lon]

                    Ui_x = Uie / 100  # m/sec
                    Ui_y = Uin / 100  # m/sec
                    Ui_z = Uiu / 100  # m/sec
                    Ui = [Ui_x, Ui_y, Ui_z]  # ExB in ENU in m/s
                    
                    ########################################################################
#                   Magnetic and Electric fields
                    ########################################################################
                    
                    # Magnetic Field from ENU to ECEF
                    pt = pyglow.Point(time_p, lat_p, lon_p, alt_p) # pyglow igrf
                    pt.run_igrf()
                    Be = pt.Bx
                    Bn = pt.By
                    Bu = pt.Bz
                    B = [Be, Bn, Bu]  # vector of magnetic field ENU in Telsa
                    Bxf[lat, lon]=Be #output (ENU)
                    Byf[lat, lon]=Bn #output (ENU)
                    Bzf[lat, lon]=Bu #output (ENU)
                    
                    Bx, By, Bz = enu_ecef(lat_p, lon_p, Be, Bn, Bu)  # Tesla in ECEF
                    B = [Be, Bn, Bu]  # vector of magnetic field ECEF in Telsa

                    # Magnetic field unit vector
                    bnorm = np.sqrt(Bx * Bx + By * By + Bz * Bz)
                    b_unit = [Bx / bnorm, By / bnorm, Bz / bnorm]
#                     bnorm = np.sqrt(Be * Be + Bn * Bn + Bu * Bu) #ENU
                    b_unit = [Be / bnorm, Bn / bnorm, Bu / bnorm] #ENU
                    
                    #Electric Field perpendicular to the magnetic field
                    Evert=-np.cross(Ui,B)  #V/m
                    Ee=Evert[0]
                    En=Evert[1]
                    Eu=Evert[2]
                    
#                     enorm=np.sqrt(En*En+Eu*Eu+Ee*Ee)
#                     e_unit=[Ee / enorm, En / enorm, Eu / enorm]
                    
#                     Evertx[lat, lon]=Ee #ENU
#                     Everty[lat, lon]=En #ENU
#                     Evertz[lat, lon]=Eu #ENU

                    #neutral wind perpendicular to the magnetic field
                    Unvert=np.cross(Un,b_unit)
                    Unvertx=Unvert[0]
                    Unverty=Unvert[1]
                    Unvertz=Unvert[2]
                    Unvertmag=np.sqrt(Unvert[0] * Unvert[0] + Unvert[1] * Unvert[1] + Unvert[2] * Unvert[2])
                    
                    unvertxf[lat, lon]=Unvertx #output
                    unvertyf[lat, lon]=Unverty #output
                    unvertzf[lat, lon]=Unvertz #output
                    
#                     TIEGCM_Electric Fields
                    Ee_TIEGCM=EEX[timer, lev, lat, lon]*100
                    En_TIEGCM=EEY[timer, lev, lat, lon]*100
                    Eu_TIEGCM=EEZ[timer, lev, lat, lon]*100
                    ETIEGCM=[Ee_TIEGCM,En_TIEGCM,Eu_TIEGCM]
                    Ee_T, En_T, Eu_T = enu_ecef(lat_p, lon_p,Ee_TIEGCM,En_TIEGCM, Eu_TIEGCM)
                    Evert2=[Ee_T,En_T,Eu_T]
                    
                    E_T_vert=np.cross(Evert,b_unit)
                    Ee_T_vert=E_T_vert[0]
                    En_T_vert=E_T_vert[1]
                    Eu_T_vert=E_T_vert[2]
                                        
                    
                    #Perpendicular electric field in neutral atmosphere reference frame                                        
                    Estar=Evert+np.cross(Unvert,B)
                                                           
                    Evertx[lat, lon]=Ee_TIEGCM #ENU
                    Everty[lat, lon]=En_TIEGCM  #ENU
                    Evertz[lat, lon]=Eu_TIEGCM  #ENU


                    #Calculate ExB velocities 
                    ExBcalc=np.cross(E_T_vert,b_unit)/(bnorm)
                    ExBx_calc[lat, lon]=ExBcalc[0]
                    ExBy_calc[lat, lon]=ExBcalc[1]
                    ExBz_calc[lat, lon]=ExBcalc[2]
                    
                    ########################################################################
#                   Gyrofrequencies and collision frequencies
                    ########################################################################

                    #gyrofrequencies (Hz)
                    bgauss=bnorm*10000
                    qeomeo10=1.7588028*10**7
                    qeoNa010=9.6489*10**3
                    rmassinv_op=1/mO
                    rmassinv_nop=1/mNO
                    rmassinv_o2p=1/mO2 
                    
                    omega_op=qeoNa010*bgauss*rmassinv_op
                    omega_o2p=qeoNa010*bgauss*rmassinv_o2p
                    omega_nop=qeoNa010*bgauss*rmassinv_nop
                    omega_op_inv=1/omega_op
                    omega_o2p_inv=1/omega_o2p
                    omega_nop_inv=1/omega_nop
                    omega_e=qeomeo10*bgauss
                    omega_e_inv=1/omega_e
                    
                    Omega_ef[lat, lon]=omega_e #output
                    Omega_Opf[lat, lon]=omega_op #output
                    Omega_O2pf[lat, lon]=omega_o2p #output
                    Omega_NOpf[lat, lon]=omega_nop #output
                    
                    #collision frequencies (Hz)
                    temps=(Ti+Tn)/2
                    rnu_o2p_o2=2.59 * (10**(-11))* np.sqrt(temps)*((1-0.073*np.log10(temps))**2)
                    rnu_op_o2=6.64 * (10**(-10))
                    rnu_nop_o2=4.27 * (10**(-10))
                    

                    
                    rnu_o2p_o=2.31*(10**(-10))
                    rnu_op_o=3.67 * (10**(-11)) * np.sqrt(temps)*((1-0.064*np.log10(temps))**2)*fcor
                    rnu_nop_o=2.44 * (10**(-10))
                    
                    rnu_o2p_n2=4.13*(10**(-10))
                    rnu_op_n2=6.82 * (10**(-10))
                    rnu_nop_n2=4.34*(10**(-10))
                    
                    rnu_o2p=(rnu_o2p_o2*NO2+rnu_o2p_o*NO+rnu_o2p_n2*NN2)*omega_o2p_inv
                    rnu_op=(rnu_op_o2*NO2+rnu_op_o*NO+rnu_op_n2*NN2)*omega_op_inv
                    rnu_nop=(rnu_nop_o2*NO2+rnu_nop_o*NO+rnu_nop_n2*NN2)*omega_nop_inv
                    
                    vOp=rnu_op_o2*NO2+rnu_op_o*NO+rnu_op_n2*NN2
                    vO2p=rnu_o2p_o2*NO2+rnu_o2p_o*NO+rnu_o2p_n2*NN2
                    vNOp=rnu_nop_o2*NO2+rnu_nop_o*NO+rnu_nop_n2*NN2
                    vin=vOp+vO2p+vNOp
                    
                    vop_o=rnu_op_o*NO
                    vop_o2=rnu_op_o2*NO2
                    vop_n2=rnu_op_n2*NN2
                    vo2p_o=rnu_o2p_o*NO
                    vo2p_o2=rnu_o2p_o2*NO2
                    vo2p_n2=rnu_o2p_n2*NN2
                    vnop_o=rnu_nop_o*NO
                    vnop_o2=rnu_nop_o2*NO2
                    vnop_n2=rnu_nop_n2*NN2
                    
                    ven=(2.33 * (10**(-11))*NN2*Te*(1-(1.21*(10**(-4)*Te))) \
                        + 1.82 * (10**(-10))*NO2*np.sqrt(Te)*(1+(3.6*(10**(-2)*np.sqrt(Te)))) \
                        +8.9 * (10**(-11))*NO*np.sqrt(Te)*(1+(5.7*(10**(-4)*Te))))
                    
                    rnu_ne=(2.33 * (10**(-11))*NN2*Te*(1-(1.21*(10**(-4)*Te))) \
                        + 1.82 * (10**(-10))*NO2*np.sqrt(Te)*(1+(3.6*(10**(-2)*np.sqrt(Te)))) \
                        +8.9 * (10**(-11))*NO*np.sqrt(Te)*(1+(5.7*(10**(-4)*Te))))*omega_e_inv
                    rnu_ne=4*rnu_ne
                    
                    vOpf[lat, lon]=vOp #Output file
                    venf[lat, lon]=ven #Output file
                    vO2pf[lat, lon]=vO2p #Output file
                    vNOpf[lat,lon]=vNOp 
                    vinf[lat, lon]=vin #Output file
                    
                    ##########################################################################
#                   Ion-electron (Coulomb) collision frequencies
                    ##########################################################################                    
                    ve_op=(54.5*NOp)/(Te**(1.5))
                    ve_o2p=(54.5*NO2p)/(Te**(1.5))
                    ve_nop=(54.5*NNOp)/(Te**(1.5))
                    ve_np=(54.5*Nplus)/(Te**(1.5))
                    ve_i=(ve_op+ve_o2p+ve_nop+ve_np)/4
                    # veif[lat,lon]=ve_i

                    ven=(2.33 * (10**(-11))*NN2*Te*(1-(1.21*(10**(-4)*Te))) \
                        + 1.82 * (10**(-10))*NO2*np.sqrt(Te)*(1+(3.6*(10**(-2)*np.sqrt(Te)))) \
                        +8.9 * (10**(-11))*NO*np.sqrt(Te)*(1+(5.7*(10**(-4)*Te))))
                    
                    ve_n2=2.33 * (10**(-11))*NN2*Te*(1-(1.21*(10**(-4)*Te)))
                    ve_o=8.9 * (10**(-11))*NO*np.sqrt(Te)*(1+(5.7*(10**(-4)*Te)))
                    ve_o2=1.82 * (10**(-10))*NO2*np.sqrt(Te)*(1+(3.6*(10**(-2)*np.sqrt(Te))))
#                     ve_he=4.6* (10**(-10))*he_cm3*np.sqrt(Te)
                    

                    
                    ########################################################################
#                   Conductivities
                    ########################################################################
    
                    qe_fac=(electron*10**10)/bgauss
                    Necm2=NOp+NO2p+NNOp
                    sigmaPed=qe_fac*((NOp*rnu_op/(1+rnu_op**2))+(NO2p*rnu_o2p/(1+rnu_o2p**2)) \
                                    +(NNOp*rnu_nop/(1+rnu_nop**2))+Necm2*rnu_ne/(1+rnu_ne**2))
                    
                    sigmaHall=qe_fac*(Necm2/(1+rnu_ne**2)-NOp/(1+rnu_op**2)-NO2p/(1+rnu_o2p**2)-NNOp/(1+rnu_nop**2))
                    
                    sigmaPedf[lat, lon]=sigmaPed*10**(5) #output 10^-5 Si/m
                    sigmaHallf[lat, lon]=sigmaHall*10**(5) #output 10^-5 Si/m

                    #Parallel conductivity
                    sigma0=(Ne*electron*electron)/(me*ven)*10**(5) #output 10^-5 Si/m
                    sigma_par[lat, lon]=sigma0
                    
                    #TIEGCM conductivities
                    sP_tiegcm=PED_all[timer, lev, lat, lon]
                    sH_tiegcm=HALL_all[timer, lev, lat, lon]
                    sigmaPedf_T[lat, lon]=sP_tiegcm*10**(5) #output 10^-5 Si/m
                    sigmaHallf_T[lat, lon]=sH_tiegcm*10**(5) #output 10^-5 Si/m
                    
                    ########################################################################
#                   Velocities
                    ########################################################################

                    #Ion Velocity in neutral reference frame (formula derived through ion momentum equation)
                    #O+
                    vi_op_star=(vOp*omega_op*Estar+omega_op**2*(np.cross(Estar,b_unit)))/(bnorm*(vOp**2+omega_op**2))
                    vi_op_starx=vi_op_star[0] #m/s
                    vi_op_stary=vi_op_star[1] #m/s
                    vi_op_starz=vi_op_star[2] #m/s
                    vi_op_starmag=np.sqrt(vi_op_star[0] * vi_op_star[0] + vi_op_star[1] * vi_op_star[1] + vi_op_star[2] * vi_op_star[2])

                    vie_op=vi_op_starx+Unvertx #from neutral frame to ENU
                    vin_op=vi_op_stary+Unverty #from neutral frame to ENU
                    viu_op=vi_op_starz+Unvertz #from neutral frame to ENU

                                       
                    vinorm_op=np.sqrt(vie_op*vie_op+vin_op*vin_op+viu_op*viu_op)
                    e_unit_op=[vie_op / vinorm_op, vin_op / vinorm_op, viu_op / vinorm_op]
                    
                    vi_mag_op[lat, lon]=vinorm_op
                    vi_starxf_op[lat, lon]=vie_op #output
                    vi_staryf_op[lat, lon]=vin_op #output
                    vi_starzf_op[lat, lon]=viu_op #output
                    
                    #O2+
                    vi_o2p_star=(vO2p*omega_o2p*Estar+omega_o2p**2*(np.cross(Estar,b_unit)))/(bnorm*(vO2p**2+omega_o2p**2))
                    vi_o2p_starx=vi_o2p_star[0] #m/s
                    vi_o2p_stary=vi_o2p_star[1] #m/s
                    vi_o2p_starz=vi_o2p_star[2] #m/s
                    vi_o2p_starmag=np.sqrt(vi_o2p_star[0] * vi_o2p_star[0] + vi_o2p_star[1] * vi_o2p_star[1] + vi_o2p_star[2] * vi_o2p_star[2])

                    vie_o2p=vi_o2p_starx+Unvertx #from neutral frame to ENU
                    vin_o2p=vi_o2p_stary+Unverty #from neutral frame to ENU
                    viu_o2p=vi_o2p_starz+Unvertz #from neutral frame to ENU

                                       
                    vinorm_o2p=np.sqrt(vie_o2p*vie_o2p+vin_o2p*vin_o2p+viu_o2p*viu_o2p)
                    e_unit_o2p=[vie_o2p / vinorm_o2p, vin_o2p / vinorm_o2p, viu_o2p / vinorm_o2p]
                    
                    vi_mag_o2p[lat, lon]=vinorm_o2p
                    vi_starxf_o2p[lat, lon]=vie_o2p #output
                    vi_staryf_o2p[lat, lon]=vin_o2p #output
                    vi_starzf_o2p[lat, lon]=viu_o2p #output
                    
                    #NO+
                    vi_nop_star=(vNOp*omega_nop*Estar+omega_nop**2*(np.cross(Estar,b_unit)))/(bnorm*(vNOp**2+omega_nop**2))
                    vi_nop_starx=vi_nop_star[0] #m/s
                    vi_nop_stary=vi_nop_star[1] #m/s
                    vi_nop_starz=vi_nop_star[2] #m/s
                    vi_nop_starmag=np.sqrt(vi_nop_star[0] * vi_nop_star[0] + vi_nop_star[1] * vi_nop_star[1] + vi_nop_star[2] * vi_nop_star[2])

                    vie_nop=vi_nop_starx+Unvertx #from neutral frame to ENU
                    vin_nop=vi_nop_stary+Unverty #from neutral frame to ENU
                    viu_nop=vi_nop_starz+Unvertz #from neutral frame to ENU

                                       
                    vinorm_nop=np.sqrt(vie_nop*vie_nop+vin_nop*vin_nop+viu_nop*viu_nop)
                    e_unit_nop=[vie_nop / vinorm_nop, vin_nop / vinorm_nop, viu_nop / vinorm_nop]
                    
                    vi_mag_nop[lat, lon]=vinorm_nop
                    vi_starxf_nop[lat, lon]=vie_nop #output
                    vi_staryf_nop[lat, lon]=vin_nop #output
                    vi_starzf_nop[lat, lon]=viu_nop #output
                    
                    #Electron Velocity in neutral reference frame (formula derived through electron momentum equation)
                    ve_star=(-ven*omega_e*Estar+omega_e**2*(np.cross(Estar,b_unit)))/(bnorm*(ven**2+omega_e**2))
                    ve_starx=ve_star[0] #m/s
                    ve_stary=ve_star[1] #m/s
                    ve_starz=ve_star[2] #m/s
                    ve_starmag=np.sqrt(ve_star[0] * ve_star[0] + ve_star[1] * ve_star[1] + ve_star[2] * ve_star[2])

                    ve_xtemp=ve_starx+Unvertx #from neutral frame to ENU
                    ve_ytemp=ve_stary+Unverty #from neutral frame to ENU
                    ve_ztemp=ve_starz+Unvertz #from neutral frame to ENU
                  
                    vee_out[lat, lon]=ve_xtemp
                    ven_out[lat, lon]=ve_ytemp
                    veu_out[lat, lon]=ve_ztemp    
                    
                    ########################################################################
#                     Magnetic parameters
                    ########################################################################
    
                    dipmag=np.arctan(Bz/np.sqrt(By**2+Bx**2))
                    sndip=np.sin(dipmag)
                    decmag=np.arctan2(By,Bx)
                    sndec=np.sin(decmag)
                    sn2dec=sndec**2
                    csdec=np.cos(decmag)
                    tm1=sn2dec+(1-sn2dec)*sndip**2
                    tm2=(1-sn2dec)+sn2dec*sndip**2
                    tm3=sndec*csdec*(1-sndip**2)
                    
                    ############################rnu##############################################
#                   Ohmic Heating 
                    ########################################################################## 
                                        
                    A=E_T_vert+np.cross(Unvert,B)
#                     L=np.sqrt(A[0] * A[0] + A[1] * A[1] + A[2] * A[2])

                    Ohmic2=sP_tiegcm*np.dot(A,A)
#                     Ohmic2=sigmaPed*L*L
                    Ohmic[lat, lon]=Ohmic2*1000 #mW/m^3
    
                    ##########################################################################
#                     Joule Heating - Frictional Heating
                    ##########################################################################
    
                    Joule_op=electron*NOp*(10**6)*np.dot(vi_op_star,A)
                    Joule_o2p=electron*NO2p*(10**6)*np.dot(vi_o2p_star,A)
                    Joule_nop=electron*NNOp*(10**6)*np.dot(vi_nop_star,A)
                    Joule1=Joule_op+Joule_o2p+Joule_nop
                    Joule[lat, lon]=Joule1*1000 #mW/m^3
                    
                    Frictional_op=MO*NOp*(10**6)*vOp*vi_op_starmag*vi_op_starmag
                    Frictional_o2p=MO2*NO2p*(10**6)*vO2p*vi_o2p_starmag*vi_o2p_starmag
                    Frictional_nop=MNO*NNOp*(10**6)*vNOp*vi_nop_starmag*vi_nop_starmag
                    Frictional1=Frictional_op+Frictional_o2p+Frictional_nop
                    Frictional[lat, lon]=Frictional1*1000 #mW/m^3
                    
                    rho=den_all[timer, lev, lat, lon]*1000
                    JH_rho[lat, lon]=(Ohmic2/rho)*1000 #mW/kg
                    
                    ##########################################################################
#                   Convection Heating - Wind Correction
                    ##########################################################################
                                        
                    UNxB=np.cross(Unvert,B)
                    UNxBmag=np.sqrt(UNxB[0] * UNxB[0] + UNxB[1] * UNxB[1] + UNxB[2] * UNxB[2])
                    ExB=np.cross(E_T_vert,B)


                    Convenction_heating=sigmaPed*np.dot(E_T_vert,E_T_vert)
                    Convenction_heating_Out[lat, lon]=Convenction_heating*1000 #mW/m^3
                    
                    conv2=sigmaPed*(tm1*Ui_x*Ui_x+tm2*Ui_y*Ui_y+2*tm3*Ui_x*Ui_y)
                    conv2=(conv2*bnorm**2)/(sndip*sndip)
                    
#                     Wind_heating=sigmaPed*UNxBmag*UNxBmag-2*sigmaPed*np.dot(Un,ExB)
                    Wind_heating=(sP_tiegcm*bnorm**2*(tm2*Un_x**2+tm1*Un_y**2-2*tm3*Un_x*Un_y
                                                      -2*(Un_x*Ui_x+Un_y*Ui_y)))
                    Wind_heating_Out[lat, lon]=Wind_heating*1000 #mW/m^3
        
                    #TIEGCM QWIND and QAMIE
                    QAMIE1=QAMIE[timer, lev, lat, lon]
                    QAMIE_T[lat,lon]=QAMIE1
                    
                    print(Convenction_heating*1000,QAMIE1,conv2*1000)
                    print(Joule1*1000,Frictional1*1000,Ohmic2*1000)
                    QWIND1=QWIND[timer, lev, lat, lon]
                    QWIND_T[lat,lon]=QWIND1                               
#                     print(Wind_heating*1000,QWIND1)
                    Ohmic_heating_OhmsLaw=Convenction_heating+Wind_heating
                    Ohmic_heating_OhmsLaw_Out[lat, lon]=Ohmic_heating_OhmsLaw*1000 #mW/m^3
#                     print(Ohmic2,Ohmic_heating_OhmsLaw)
                    
                    ##########################################################################
#                   Ohmic per Pressure/ Pressure per Ohmic
                    ##########################################################################
                                   
                    Ohmic_pressure=(Ohmic2/pressuref_Pa)
                    Ohmic_Press[lat,lon]=Ohmic_pressure
                
                    pressure_ohmic_tmp=(1/Ohmic_pressure)
                    Press_Ohmic[lat,lon]=pressure_ohmic_tmp
                    
                    ##########################################################################
#                   Electric Currents
                    ##########################################################################
                    
                    #perpendicular electric current density (A/m^2) (Ohm's LaW)
                    J_p_Ohmic=sigmaPed*Estar+sigmaHall*np.cross(b_unit,Estar)
                    J_p_Ohmicx=J_p_Ohmic[0]
                    J_p_Ohmicy=J_p_Ohmic[1]
                    J_p_Ohmicz=J_p_Ohmic[2]
                    
                    #Pedersen and Hall currents
                    J_Pedersen=sigmaPed*Estar
                    J_Hall=sigmaHall*np.cross(b_unit,Estar)
                    
                    Jpe=J_Pedersen[0]
                    Jpn=J_Pedersen[1]
                    Jpu=J_Pedersen[2]
                    Jhe=J_Hall[0]
                    Jhn=J_Hall[1]
                    Jhu=J_Hall[2]

                    
                    #####################
                    Jpnorm=np.sqrt(Jpn*Jpn+Jpu*Jpu+Jpe*Jpe)
                    Jp_unit=[Jpe / Jpnorm, Jpn / Jpnorm, Jpu / Jpnorm]
                    
                    Jhnorm=np.sqrt(Jhn*Jhn+Jhu*Jhu+Jhe*Jhe)
                    Jh_unit=[Jhe / Jhnorm, Jhn / Jhnorm, Jhu / Jhnorm]
                    
                    J_Pede[lat,lon]=Jpe #output 
                    J_Pedn[lat,lon]=Jpn #output  
                    J_Pedu[lat,lon]=Jpu #output 
                    J_Halle[lat,lon]=Jhe #output 
                    J_Halln[lat,lon]=Jhn #output 
                    J_Hallu[lat,lon]=Jhu #output 
                    J_Ohme[lat, lon]=Jpe+Jhe #output 
                    J_Ohmn[lat, lon]=Jpn+Jhn #output 
                    J_Ohmu[lat, lon]=Jpu+Jhu #output 
                    
                    J_mag[lat,lon]=np.sqrt((Jpe+Jhe)**2+(Jpn+Jhn)**2+(Jpu+Jhu)**2) #output 

                    #perpendicular electric current density (A/m^2) (densities)
                    J_den=electron*Ne*(vi_op_star-ve_star)
                    J_denx=J_den[0]
                    J_deny=J_den[1]
                    J_denz=J_den[2]
                    
                    ##########################################################################
#                   JxB forcing
                    ##########################################################################

                    JxB=np.cross(J_p_Ohmic,B)
                    JxBe[lat, lon]=JxB[0]
                    JxBn[lat, lon]=JxB[1]
                    JxBu[lat, lon]=JxB[2]
                    
                    ##########################################################################
#                   Mechanical Power and Energy
                    ##########################################################################
                    
                    Mechanical_power=np.dot(Un,JxB)
                    Mechanical_power_Out[lat, lon]=Mechanical_power
                    Energy=Ohmic_heating_OhmsLaw+Mechanical_power
                    Energy_Out[lat, lon]=Energy
                    
                    ##########################################################################
#                   Cross Sections
                    ##########################################################################
                    N_neutral=(NO2+NN2+NO)*10**6
                    MN2=(mN2)/(NA*1000)
                    M_ion=MO+MO2+MNO
                    
                    cross=(vin/N_neutral)/(np.sqrt((2*boltzmann*Ti)/M_ion))
                    cross_section[lat,lon]=cross
                    
                    cross_op=(vOp/N_neutral)/(np.sqrt((2*boltzmann*Ti)/MO))
                    cross_section_op[lat,lon]=cross_op
                    
                    cross_o2p=(vO2p/N_neutral)/(np.sqrt((2*boltzmann*Ti)/MO2))
                    cross_section_o2p[lat,lon]=cross_o2p
                    
                    cross_nop=(vNOp/N_neutral)/(np.sqrt((2*boltzmann*Ti)/MNO))
                    cross_section_nop[lat,lon]=cross_nop
                    
                    warnings.simplefilter('ignore')
                    
                    #print(Ohmic2*1000,Ohmic_heating_OhmsLaw*1000,Convenction_heating*1000,Wind_heating*1000 )
                    
                    
                    ##########################################################################
#                     Ion heating rates
                    ##########################################################################
    
                    qDTop_o=3*NOp*(10 ** 6)*boltzmann_si*vop_o*(MO/(MO+MO))*(Ti-Tn)
                    qDTop_o2=3*NOp*(10 ** 6)*boltzmann_si*vop_o2*(MO/(MO2+MO))*(Ti-Tn)
                    qDTop_n2=3*NOp*(10 ** 6)*boltzmann_si*vop_n2*(MO/(MN2+MO))*(Ti-Tn)
                    qDTo2p_o=3*NO2p*(10 ** 6)*boltzmann_si*vo2p_o*(MO2/(MO+MO2))*(Ti-Tn)
                    qDTo2p_o2=3*NO2p*(10 ** 6)*boltzmann_si*vo2p_o2*(MO2/(MO2+MO2))*(Ti-Tn)
                    qDTo2p_n2=3*NO2p*(10 ** 6)*boltzmann_si*vo2p_n2*(MO2/(MN2+MO2))*(Ti-Tn)
                    qDTnop_o=3*NNOp*(10 ** 6)*boltzmann_si*vnop_o*(MNO/(MO+MNO))*(Ti-Tn)
                    qDTnop_o2=3*NNOp*(10 ** 6)*boltzmann_si*vnop_o2*(MNO/(MO2+MNO))*(Ti-Tn)
                    qDTnop_n2=3*NNOp*(10 ** 6)*boltzmann_si*vnop_n2*(MNO/(MN2+MNO))*(Ti-Tn)
                    
                    qDTop_n=qDTop_o+qDTop_o2+qDTop_n2
                    qDTo2p_n=qDTo2p_o+qDTo2p_o2+qDTo2p_n2
                    qDTnop_n=qDTnop_o+qDTnop_o2+qDTnop_n2
                    
                    qDTi_n=qDTop_n+qDTo2p_n+qDTnop_n
                    
                    qDTop_of[lat,lon]=qDTop_o
                    qDTop_o2f[lat,lon]=qDTop_o2
                    qDTop_n2f[lat,lon]=qDTop_n2
                    qDTo2p_of[lat,lon]=qDTo2p_o
                    qDTo2p_o2f[lat,lon]=qDTo2p_o2
                    qDTo2p_n2f[lat,lon]=qDTo2p_n2
                    qDTnop_of[lat,lon]=qDTnop_o
                    qDTnop_o2f[lat,lon]=qDTnop_o2
                    qDTnop_n2f[lat,lon]=qDTnop_n2
                    
                    qDTop_nf[lat,lon]=qDTop_n
                    qDTo2p_nf[lat,lon]=qDTo2p_n
                    qDTnop_nf[lat,lon]=qDTnop_n
                    
                    qDTi_nf[lat,lon]=qDTi_n*1000 #mW/m^3
                    
                    ##########################################################################
#                     Cooling rates
                    ##########################################################################
                    
                    #N2 rotation
                    Le_N2_rotf=((3.5*10**(-14)*Necm*NN2*(Te-Tn))/np.sqrt(Te))*1.60217662*10**(-16)
                    Le_N2_rot_2d[lat,lon]=Le_N2_rotf
                    #O2 rotation
                    Le_O2_rotf=((5.2*10**(-15)*Necm*NO2*(Te-Tn))/np.sqrt(Te))*1.60217662*10**(-16)
                    Le_O2_rot_2d[lat,lon]=Le_O2_rotf
                    #O2 vibration
                    logQ_Te=-19.9171+0.0267*Te-3.9960*(10**(-5))*(Te**2) \
                            +3.5187*(10**(-8))*(Te**3)-1.9228*(10**(-11))*(Te**4) \
                            +6.6865*(10**(-15))*(Te**5)-1.4791*(10**(-18))*(Te**6) \
                            +2.0127*(10**(-22))*(Te**7)-1.5346*(10**(-26))*(Te**8) \
                            +5.0148*(10**(-31))*(Te**9)

                    Q_Te=10**(logQ_Te)


                    Le_O2_vibf=(Necm*NO2*Q_Te*(1-np.exp(2239*(Te**(-1)-Tn**(-1)))))*1.60217662*10**(-16) 
                    Le_O2_vib_2d[lat,lon]=Le_O2_vibf
                    
                    #O fine structure
                    D=5+np.exp(-326.6/Tn)+3*np.exp(-227.7/Tn)
                    S21=1.863*10**(-11)
                    S20=1.191*10**(-11)
                    S10=8.249*10**(-16)*Te**(0.6)*np.exp(-227.7/Tn)
                    
                    Le_O_finef=(((Necm*NO)/D)*((S10*(1-np.exp(98.9*(1/Te-1/Tn)))) \
                                               +S20*(1-np.exp(326.6*(1/Te-1/Tn))) \
                                               +S21*(1-np.exp(227.7*(1/Te-1/Tn)))))*1.60217662*10**(-16) 
                    
                    Le_O_fine_2d[lat,lon]=Le_O_finef
                    
                    #N2 vibration
                    E1=3353 #K
                    Tvib=Tn

                    A01=2.025
                    A02=-7.066
                    A03=-8.211
                    A04=-9.713
                    A05=-10.353
                    A06=-10.819
                    A07=-10.183
                    A08=-12.698
                    A09=-14.710
                    A010=-17.538
                    B01=8.782*10**(-4)
                    B02=1.0001*10**(-2)
                    B03=1.092*10**(-2)
                    B04=1.204*10**(-2)
                    B05=1.243*10**(-2)
                    B06=1.244*10**(-2)
                    B07=1.185*10**(-2)
                    B08=1.309*10**(-2)
                    B09=1.409*10**(-2)
                    B010=1.6*10**(-2)
                    C01=2.954*10**(-7)
                    C02=-3.066*10**(-6)
                    C03=-3.369*10**(-6)
                    C04=-3.732*10**(-6)
                    C05=-3.850*10**(-6)
                    C06=-3.771*10**(-6)
                    C07=-3.570*10**(-6)
                    C08=-3.952*10**(-6)
                    C09=-4.249*10**(-6)
                    C010=-4.916*10**(-6)
                    D01=-9.562*10**(-11)
                    D02=4.436*10**(-10)
                    D03=4.891*10**(-10)
                    D04=5.431*10**(-10)
                    D05=5.600*10**(-10)
                    D06=5.385*10**(-10)
                    D07=5.086*10**(-10)
                    D08=5.636*10**(-10)
                    D09=6.058*10**(-10)
                    D010=7.128*10**(-10)
                    F01=7.252*10**(-15)
                    F02=-2.449*10**(-14)
                    F03=-2.706*10**(-14)
                    F04=-3.008*10**(-14)
                    F05=-3.100*10**(-14)
                    F06=-2.936*10**(-14)
                    F07=-2.769*10**(-14)
                    F08=-3.071*10**(-14)
                    F09=-3.3*10**(-14)
                    F010=-3.941*10**(-14)

                    A12=-3.413
                    A13=-4.160
                    A14=-5.193
                    A15=-5.939
                    A16=-8.261
                    A17=-8.185
                    A18=-10.823
                    A19=-11.273
                    B12=7.326*10**(-3)
                    B13=7.803*10**(-3)
                    B14=8.360*10**(-3)
                    B15=8.807*10**(-3)
                    B16=1.01*10**(-2)
                    B17=1.01*10**(-2)
                    B18=1.199*10**(-2)
                    B19=1.283*10**(-2)
                    C12=-2.2*10**(-6)
                    C13=-2.352*10**(-6)
                    C14=-2.526*10**(-6)
                    C15=-2.669*10**(-6)
                    C16=-3.039*10**(-6)
                    C17=-3.039*10**(-6)
                    C18=-3.62*10**(-6)
                    C19=-3.879*10**(-6)
                    D12=3.128*10**(-10)
                    D13=3.352*10**(-10)
                    D14=3.606*10**(-10)
                    D15=3.806*10**(-10)
                    D16=4.318*10**(-10)
                    D17=4.318*10**(-10)
                    D18=5.159*10**(-10)
                    D19=5.534*10**(-10)
                    F12=-1.702*10**(-14)
                    F13=-1.828*10**(-14)
                    F14=-1.968*10**(-14)
                    F15=-2.073*10**(-14)
                    F16=-2.347*10**(-14)
                    F17=-2.347*10**(-14)
                    F18=-2.810*10**(-14)
                    F19=-3.016*10**(-14)


                    logQ01=A01+B01*Te+C01*Te**2+D01*Te**3+F01*Te**4-16
                    logQ02=A02+B02*Te+C02*Te**2+D02*Te**3+F02*Te**4-16
                    logQ03=A03+B03*Te+C03*Te**2+D03*Te**3+F03*Te**4-16
                    logQ04=A04+B04*Te+C04*Te**2+D04*Te**3+F04*Te**4-16
                    logQ05=A05+B05*Te+C05*Te**2+D05*Te**3+F05*Te**4-16
                    logQ06=A06+B06*Te+C06*Te**2+D06*Te**3+F06*Te**4-16
                    logQ07=A07+B07*Te+C07*Te**2+D07*Te**3+F07*Te**4-16
                    logQ08=A08+B08*Te+C08*Te**2+D08*Te**3+F08*Te**4-16
                    logQ09=A09+B09*Te+C09*Te**2+D09*Te**3+F09*Te**4-16
                    logQ010=A010+B010*Te+C010*Te**2+D010*Te**3+F010*Te**4-16

                    logQ12=A12+B12*Te+C12*Te**2+D12*Te**3+F12*Te**4-16
                    logQ13=A13+B13*Te+C13*Te**2+D13*Te**3+F13*Te**4-16
                    logQ14=A14+B14*Te+C14*Te**2+D14*Te**3+F14*Te**4-16
                    logQ15=A15+B15*Te+C15*Te**2+D15*Te**3+F15*Te**4-16
                    logQ16=A16+B16*Te+C16*Te**2+D16*Te**3+F16*Te**4-16
                    logQ17=A17+B17*Te+C17*Te**2+D17*Te**3+F17*Te**4-16
                    logQ18=A18+B18*Te+C18*Te**2+D18*Te**3+F18*Te**4-16
                    logQ19=A19+B19*Te+C19*Te**2+D19*Te**3+F19*Te**4-16

                    Le_N2_vibf=(10**(-12)*Necm*NN2*(1-np.exp(-E1/Tvib)) \
                            *((10**(logQ01)*(1-np.exp(1*E1*(Te**(-1)-Tvib**(-1))))) \
                            +(10**(logQ02)*(1-np.exp(2*E1*(Te**(-1)-Tvib**(-1)))))  \
                            +(10**(logQ03)*(1-np.exp(3*E1*(Te**(-1)-Tvib**(-1)))))  \
                            +(10**(logQ04)*(1-np.exp(4*E1*(Te**(-1)-Tvib**(-1)))))   \
                            +(10**(logQ05)*(1-np.exp(5*E1*(Te**(-1)-Tvib**(-1)))))   \
                            +(10**(logQ06)*(1-np.exp(6*E1*(Te**(-1)-Tvib**(-1)))))   \
                            +(10**(logQ07)*(1-np.exp(7*E1*(Te**(-1)-Tvib**(-1)))))   \
                            +(10**(logQ08)*(1-np.exp(8*E1*(Te**(-1)-Tvib**(-1)))))   \
                            +(10**(logQ09)*(1-np.exp(9*E1*(Te**(-1)-Tvib**(-1)))))   \
                            +(10**(logQ010)*(1-np.exp(10*E1*(Te**(-1)-Tvib**(-1)))))    \
                             )  \
                             +10**(-12)*Necm*NN2*(1-np.exp(-E1/Tvib))*np.exp(-E1/Tvib)  \
                             *((10**(logQ12)*(1-np.exp(1*E1*(Te**(-1)-Tvib**(-1)))))  \
                            +(10**(logQ13)*(1-np.exp(2*E1*(Te**(-1)-Tvib**(-1)))))  \
                            +(10**(logQ14)*(1-np.exp(3*E1*(Te**(-1)-Tvib**(-1)))))   \
                            +(10**(logQ15)*(1-np.exp(4*E1*(Te**(-1)-Tvib**(-1)))))   \
                            +(10**(logQ16)*(1-np.exp(5*E1*(Te**(-1)-Tvib**(-1)))))   \
                            +(10**(logQ17)*(1-np.exp(6*E1*(Te**(-1)-Tvib**(-1)))))   \
                            +(10**(logQ18)*(1-np.exp(7*E1*(Te**(-1)-Tvib**(-1)))))   \
                            +(10**(logQ19)*(1-np.exp(8*E1*(Te**(-1)-Tvib**(-1)))))   \
                            ))*1.60217662*10**(-16)  
                    
                    Le_N2_vib_2d[lat,lon]=Le_N2_vibf



                    #O(1D) excitation
                    d_i=2.4*10**4+0.3*(Te-1500)-1.947*(10**(-5))*(Te-1500)*(Te-4000)
                    
                    Le_O_1Df=(1.57*10**(-12)*Necm*NO*np.exp(d_i*((Te-3000)/(3000*Te))) \
                              *(np.exp(-22713*((Te-Tn)/(Te*Tn)))-1))*1.60217662*10**(-16)
                    
                    Le_O_1D_2d[lat,lon]=Le_O_1Df
                    
                    #electron cooling rates sum
                    Lef=Le_N2_rotf+Le_N2_vibf+Le_O2_vibf+Le_O_finef+Le_O_1Df+Le_O2_rotf
                    # LeOp=
                   
                    Le_all_2d[lat,lon]=Lef
                    # print(Le_N2_rotf,Le_N2_rotf,Le_O2_vibf,Le_O_finef,Le_O_1Df,Lef)
                    
                    #Electron neutral elastic collisions
                    qDTe_of=3*Necm*(10 ** 6)*boltzmann_si*ve_o*(me/(me+MO))*(Te-Tn)
                    qDTe_o2f=3*Necm*(10 ** 6)*boltzmann_si*ve_o2*(me/(me+MO2))*(Te-Tn)
                    qDTe_n2f=3*Necm*(10 ** 6)*boltzmann_si*ve_n2*(me/(me+MN2))*(Te-Tn)
#                     qDTe_hef=3*Necm*(10 ** 6)*boltzmann_si*ve_he*(me/(me+MHE))*(Te-Tn)
                    
                    qDTe_n_elastic=qDTe_of+qDTe_o2f+qDTe_n2f #+qDTe_hef
                    
                    qDTe_o_2d[lat,lon]=qDTe_of
                    qDTe_o2_2d[lat,lon]=qDTe_o2f
                    qDTe_n2_2d[lat,lon]=qDTe_n2f
#                     qDTe_he[lat,lon]=qDTe_hef
                    qDTe_n_2d[lat,lon]=qDTe_n_elastic

                    #electron-ion coulomp collisions
                    qDTe_opf=3*Necm*(10 ** 6)*boltzmann_si*ve_op*(me/(me+MO))*(Te-Ti)
                    qDTe_o2pf=3*Necm*(10 ** 6)*boltzmann_si*ve_o2p*(me/(me+MO2))*(Te-Ti)
                    qDTe_nopf=3*Necm*(10 ** 6)*boltzmann_si*ve_nop*(me/(me+MNO))*(Te-Ti)
                    qDTe_npf=3*Necm*(10 ** 6)*boltzmann_si*ve_np*(me/(me+MN))*(Te-Ti)
                    qDTe_if=qDTe_opf+qDTe_o2pf+qDTe_nopf+qDTe_npf
                    
                    qDTe_op_2d[lat,lon]=qDTe_opf
                    qDTe_o2p_2d[lat,lon]=qDTe_o2pf
                    qDTe_nop_2d[lat,lon]=qDTe_nopf
                    qDTe_np_2d[lat,lon]=qDTe_npf
                    qDTe_i_2d[lat,lon]=qDTe_if

                    
    print("Calculation executed!")
    print("Total run time: %s seconds" % (time.time() - start_time))
                    
    print("Calculation executed!")
    print("Total run time: %s seconds" % (time.time() - start_time))
tiegcm_file='../data/tiegcm_dres.s_mar2015_amie_v1_31_Rerun.nc'    
timer_value=22
pressure_level=12
    
run_map(tiegcm_file,timer_value,pressure_level)    
```
