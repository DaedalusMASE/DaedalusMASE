```python
import sys
from netCDF4 import Dataset
import numpy as np
import datetime
from ipywidgets import*
import ipywidgets as widgets
import warnings
import glob
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from mpl_toolkits.basemap import Basemap, shiftgrid, addcyclic
import matplotlib.mlab as mlab
from mpl_toolkits.axes_grid1 import make_axes_locatable
from scipy.interpolate import griddata
from pylab import *
from matplotlib.patches import Polygon
import time
import pandas as pd
from scipy.special import erf
import matplotlib.colors as colors
from matplotlib import ticker, cm
import igrf12
```

```python
#Define daedalusmase_derived_products package location
fpath = os.path.abspath("../")
print(fpath)
#add src path to python path
sys.path.append(fpath)

from daedalusmase_collision_frequencies.mod_utils import allocations as alloc
from daedalusmase_collision_frequencies.mod_utils import const
import daedalusmase_collision_frequencies.mod_utils as utils

import daedalusmase_collision_frequencies.mod_collision_freqs as cols

import daedalusmase_collision_frequencies.mod_plot_utils as plots

```

```python
def run(lat_value,lon_value,timer_value,min_alt,max_alt,d_alt):
    
    alt_range=np.arange(min_alt,max_alt,d_alt)
    for i in range(0,len(alt_range)):
        
        B_enu,b_unit_enu=utils.run_igrf(timer_value, lat_value, lon_value, alt_range[i])
        ne, Op, O2p, NOp, Np, Tn_iri, Ti, Te=utils.run_iri(timer_value, lat_value, lon_value, alt_range[i])
        O,O2,N,N2,He,H,Ar,Ntotal,Tn=utils.run_msis(timer_value, lat_value, lon_value, alt_range[i])
        
        ne_cm=ne*10**(-6)
        Op_cm=Op*10**(-6)
        O2p_cm=O2p*10**(-6)
        NOp_cm=NOp*10**(-6)
        Np_cm=Np*10**(-6)
        O_cm=O*10**(-6)
        O2_cm=O2*10**(-6)
        N_cm=N*10**(-6)
        N2_cm=N2*10**(-6)
        He_cm=He*10**(-6)
        
        alloc.ne_den.append(ne)
        alloc.Op_den.append(Op)
        alloc.O2p_den.append(O2p)
        alloc.NOp_den.append(NOp)
        alloc.Npp_den.append(Np)
        
        alloc.O_den.append(O)
        alloc.O2_den.append(O2)
        alloc.N2_den.append(N2)
       
        
        bnorm = np.sqrt(B_enu[0]*B_enu[0] + B_enu[1]*B_enu[1] + B_enu[2]*B_enu[2])
        
        ########################################################################
#                   Gyrofrequencies and collision frequencies
        ########################################################################

        #gyrofrequencies (Hz)
        
        #O+_O freqs
        
        Tr,q_dalgarno,q_banks,q_stubbe,q_sw,q_salah,q_pesnell,q_hickmann,q_shunk_nagy,q_richmond,q_ieda,q_baily_belan=cols.op_o_cols(Ti,Tn)


        alloc.q_dalgarno_iri.append(q_dalgarno)
        alloc.q_banks_iri.append(q_banks)
        alloc.q_stubbe_iri.append(q_stubbe)
        alloc.q_sw_iri.append(q_sw)
        alloc.q_salah_iri.append(q_salah)
        alloc.q_pesnell_iri.append(q_pesnell)
        alloc.q_hickmann_iri.append(q_hickmann)
        alloc.q_shunk_nagy_iri.append(q_shunk_nagy)
        alloc.q_richmond_iri.append(q_richmond)
        alloc.q_ieda_iri.append(q_ieda)
        alloc.Tr_iri.append(Tr)
        alloc.q_baily_belan_iri.append(q_baily_belan)
        #Schunk and Nagy 2009 collision frequencies (1/s)
        
        vin,vop_o,vop_o2,vop_n2,vop_he,vo2p_o,vo2p_o2,vo2p_n2,vo2p_he,vnop_o,vnop_o2,vnop_n2,vnop_he,vnp_o,vnp_o2,vnp_n2,vnp_he,vOp,vO2p,vNOp,vNp=cols.vin_schunk_and_nagy(B_enu,Ti,Tn,O2_cm,O_cm,N2_cm,He_cm)

        alloc.vOp_SN.append(vOp)
        alloc.vO2p_SN.append(vO2p)
        alloc.vNOp_SN.append(vNOp)
        alloc.vin_SN.append(vin)
        
        vinri,vinNOpr,vinO2pr,vinOpr,vinNOp_N2r,vinNOp_O2r,vinNOP_Or,vinO2p_N2r,vinO2p_O2r,vinO2p_Or,vinOp_N2r,vinOp_O2r,vinOp_Or=cols.vin_richmond(Ti,Tn,O2,O,N2)
        
        alloc.vOp_R.append(vinOpr)
        alloc.vO2p_R.append(vinO2pr)
        alloc.vNOp_R.append(vinNOpr)
        alloc.vin_R.append(vinri)  
        
        vinsw,vinNOpsw,vinO2psw,vinOpsw,vinNOp_N2sw,vinNOp_O2sw,vinNOP_Osw,vinO2p_N2sw,vinO2p_O2sw,vinO2p_Osw,vinOp_N2sw,vinOp_O2sw,vinOp_Osw=cols.vin_schunk_walker(Ti,Tn,O2_cm,O_cm,N2_cm)
        
        alloc.vOp_SW.append(vinOpsw)
        alloc.vO2p_SW.append(vinO2psw)
        alloc.vNOp_SW.append(vinNOpsw)
        alloc.vin_SW.append(vinsw)          
        
        vinb,vinNOpb,vinO2pb,vinOpb,vinNOp_N2b,vinNOp_O2b,vinNOP_Ob,vinO2p_N2b,vinO2p_O2b,vinO2p_Ob,vinOp_N2b,vinOp_O2b,vinOp_Ob=cols.vin_banks(Ti,Tn,O2,O,N2)
        
        alloc.vOp_B.append(vinOpb)
        alloc.vO2p_B.append(vinO2pb)
        alloc.vNOp_B.append(vinNOpb)
        alloc.vin_B.append(vinb)              
        
        vini,vinNOpi,vinO2pi,vinOpi,vinNOp_N2i,vinNOp_O2i,vinNOP_Oi,vinO2p_N2i,vinO2p_O2i,vinO2p_Oi,vinOp_N2i,vinOp_O2i,vinOp_Oi=cols.vin_ieda(Ti,Tn,O2,O,N2)
        
        alloc.vOp_I.append(vinOpi)
        alloc.vO2p_I.append(vinO2pi)
        alloc.vNOp_I.append(vinNOpi)
        alloc.vin_I.append(vini) 
        
        ven, ve_n2, ve_o, ve_o2=cols.ven_coll_freq(B_enu,N2_cm,O2_cm,O_cm,Te)

        sigmaped_sn,sigmaped_e_sn,sigmaped_op_sn,sigmaped_o2p_sn,sigmaped_nop_sn=cols.pedersen(bnorm,ne,Op,O2p,NOp,ven,vOp,vO2p,vNOp)
        sigmaped_r,sigmaped_e_r,sigmaped_op_r,sigmaped_o2p_r,sigmaped_nop_r=cols.pedersen(bnorm,ne,Op,O2p,NOp,ven,vinOpr,vinO2pr,vinNOpr)
        sigmaped_sw,sigmaped_e_sw,sigmaped_op_sw,sigmaped_o2p_sw,sigmaped_nop_sw=cols.pedersen(bnorm,ne,Op,O2p,NOp,ven,vinOpsw,vinO2psw,vinNOpsw)
        sigmaped_b,sigmaped_e_b,sigmaped_op_b,sigmaped_o2p_b,sigmaped_nop_b=cols.pedersen(bnorm,ne,Op,O2p,NOp,ven,vinOpb,vinO2pb,vinNOpb)
        sigmaped_i,sigmaped_e_i,sigmaped_op_i,sigmaped_o2p_i,sigmaped_nop_i=cols.pedersen(bnorm,ne,Op,O2p,NOp,ven,vinOpi,vinO2pi,vinNOpi)        

        alloc.sigmaped_SN.append(sigmaped_sn)
        alloc.sigmaped_e_SN.append(sigmaped_e_sn)
        alloc.sigmaped_op_SN.append(sigmaped_op_sn)
        alloc.sigmaped_o2p_SN.append(sigmaped_o2p_sn)
        alloc.sigmaped_nop_SN.append(sigmaped_nop_sn)

        alloc.sigmaped_R.append(sigmaped_r)
        alloc.sigmaped_e_R.append(sigmaped_e_r)
        alloc.sigmaped_op_R.append(sigmaped_op_r)
        alloc.sigmaped_o2p_R.append(sigmaped_o2p_r)
        alloc.sigmaped_nop_R.append(sigmaped_nop_r)
        
        alloc.sigmaped_SW.append(sigmaped_sw)
        alloc.sigmaped_e_SW.append(sigmaped_e_sw)
        alloc.sigmaped_op_SW.append(sigmaped_op_sw)
        alloc.sigmaped_o2p_SW.append(sigmaped_o2p_sw)
        alloc.sigmaped_nop_SW.append(sigmaped_nop_sw)        
        
        alloc.sigmaped_B.append(sigmaped_b)
        alloc.sigmaped_e_B.append(sigmaped_e_b)
        alloc.sigmaped_op_B.append(sigmaped_op_b)
        alloc.sigmaped_o2p_B.append(sigmaped_o2p_b)
        alloc.sigmaped_nop_B.append(sigmaped_nop_b)  
        
        alloc.sigmaped_I.append(sigmaped_i)
        alloc.sigmaped_e_I.append(sigmaped_e_i)
        alloc.sigmaped_op_I.append(sigmaped_op_i)
        alloc.sigmaped_o2p_I.append(sigmaped_o2p_i)
        alloc.sigmaped_nop_I.append(sigmaped_nop_i)  
        
        sigmahall_sn,sigmahall_e_sn,sigmahall_op_sn,sigmahall_o2p_sn,sigmahall_nop_sn=cols.hall(bnorm,ne,Op,O2p,NOp,ven,vOp,vO2p,vNOp)
        sigmahall_r,sigmahall_e_r,sigmahall_op_r,sigmahall_o2p_r,sigmahall_nop_r=cols.hall(bnorm,ne,Op,O2p,NOp,ven,vinOpr,vinO2pr,vinNOpr)
        sigmahall_sw,sigmahall_e_sw,sigmahall_op_sw,sigmahall_o2p_sw,sigmahall_nop_sw=cols.hall(bnorm,ne,Op,O2p,NOp,ven,vinOpsw,vinO2psw,vinNOpsw)
        sigmahall_b,sigmahall_e_b,sigmahall_op_b,sigmahall_o2p_b,sigmahall_nop_b=cols.hall(bnorm,ne,Op,O2p,NOp,ven,vinOpb,vinO2pb,vinNOpb)
        sigmahall_i,sigmahall_e_i,sigmahall_op_i,sigmahall_o2p_i,sigmahall_nop_i=cols.hall(bnorm,ne,Op,O2p,NOp,ven,vinOpi,vinO2pi,vinNOpi)        

        
        alloc.sigmahall_SN.append(sigmahall_sn)
        alloc.sigmahall_e_SN.append(sigmahall_e_sn)
        alloc.sigmahall_op_SN.append(sigmahall_op_sn)
        alloc.sigmahall_o2p_SN.append(sigmahall_o2p_sn)
        alloc.sigmahall_nop_SN.append(sigmahall_nop_sn)

        alloc.sigmahall_R.append(sigmahall_r)
        alloc.sigmahall_e_R.append(sigmahall_e_r)
        alloc.sigmahall_op_R.append(sigmahall_op_r)
        alloc.sigmahall_o2p_R.append(sigmahall_o2p_r)
        alloc.sigmahall_nop_R.append(sigmahall_nop_r)
        
        alloc.sigmahall_SW.append(sigmahall_sw)
        alloc.sigmahall_e_SW.append(sigmahall_e_sw)
        alloc.sigmahall_op_SW.append(sigmahall_op_sw)
        alloc.sigmahall_o2p_SW.append(sigmahall_o2p_sw)
        alloc.sigmahall_nop_SW.append(sigmahall_nop_sw)        
        
        alloc.sigmahall_B.append(sigmahall_b)
        alloc.sigmahall_e_B.append(sigmahall_e_b)
        alloc.sigmahall_op_B.append(sigmahall_op_b)
        alloc.sigmahall_o2p_B.append(sigmahall_o2p_b)
        alloc.sigmahall_nop_B.append(sigmahall_nop_b)  
        
        alloc.sigmahall_I.append(sigmahall_i)
        alloc.sigmahall_e_I.append(sigmahall_e_i)
        alloc.sigmahall_op_I.append(sigmahall_op_i)
        alloc.sigmahall_o2p_I.append(sigmahall_o2p_i)
        alloc.sigmahall_nop_I.append(sigmahall_nop_i)         
```

## Define simulation parameters

```python
#simulation time
time_sim=datetime.datetime(2012, 3, 17, 0, 0, 0)
#altiltude resolution in km
dalt_sim=1
#longitude in deg
lon_sim=-117.5
#latitude in deg
lat_sim=36.5
#minimum altitude in km
altmin_sim=100
#maximum altitude in km
altmax_sim=500
#minimum temperature in K
Tmin=500
#maximum temperature in K
Tmax=2000
#temperature resolution in k
dT=10
#run simulation
run(lat_sim,lon_sim,time_sim,altmin_sim,altmax_sim,dalt_sim)

```
